## What is Matplotlib?

Matplotlib is Python's most popular graph library.

Import it: **import matplotlib.pyplot as plt**

Think: plt = pencil

We use it to draw.

### First Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)

plt.show()
```

Imagine points: (1,10), (2,20)
(3,30)
(4,40)

Python joins them with lines.

A line graph appears.

## Anatomy of a Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)

plt.title("My First Graph")

plt.xlabel("X Values")

plt.ylabel("Y Values")

plt.show()
```

title()  → Graph name

xlabel() → Bottom label

ylabel() → Side label

show()   → Display graph

### Real Student Example

```python
import matplotlib.pyplot as plt

students = ["Ali","Sara","John","Mary"]

marks = [85,90,78,95]

plt.plot(students, marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

Graph shows marks visually.

## Line Chart

A line chart shows:

Change over time

Examples:\
Stock prices\
Temperature\
Sales growth\
Website visitors

Example:

```python
days = [1,2,3,4,5]

temperature = [25,27,29,31,30]

plt.plot(days, temperature)

plt.show()
```

## Multiple Lines

Compare two students.

```python
import matplotlib.pyplot as plt

tests = [1,2,3,4]

ali = [70,75,80,85]

sara = [80,82,85,90]

plt.plot(tests, ali)

plt.plot(tests, sara)

plt.show()
```

## Grid

Grid makes graphs easier to read.

```python
plt.grid()
```

```python
plt.plot(x,y)

plt.grid()

plt.show()
```

## Markers

Show points clearly.

```python
plt.plot(x,y, marker="o")
```

## Different Line Styles

Dashed line:

```python
plt.plot(x,y, linestyle="--")
```

Dotted line:

```python
plt.plot(x,y, linestyle=":")
```

**Example**

Monthly Sales:

```python
import matplotlib.pyplot as plt

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    1000,
    1500,
    2500,
    4000
]

plt.plot(months, sales)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid()

plt.show()
```

Question: Which month had highest sales?

Easy to see.

## Mini Project

Student Marks Visualization

```python
import matplotlib.pyplot as plt

students = [
    "Ali",
    "Sara",
    "John",
    "Mary"
]

marks = [
    85,
    90,
    78,
    95
]

plt.plot(
    students,
    marks,
    marker="o"
)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.grid()

plt.show()
```

**plt.plot()** Means **Draw a line graph**
    
**plt.title()** Means **Name of graph**

**plt.xlabel()** means **Bottom label**

**plt.ylabel()** means **Side label**

**plt.show()** means **Display graph**
