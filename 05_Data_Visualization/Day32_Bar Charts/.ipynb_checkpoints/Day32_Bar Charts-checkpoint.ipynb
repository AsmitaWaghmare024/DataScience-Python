# Bar Charts

**plt.bar()** which creates a **Bar Chart**.

What is a Bar Chart?

Imagine 4 students:

Ali   85
Sara  90
John  78
Mary  95

Instead of showing numbers:

Ali   ████████████████

Sara  ██████████████████

John  ██████████████

Mary  ███████████████████

Each bar represents a value.

This is called a **Bar Chart**

#### Why Companies Love Bar Charts

Suppose a company sells:

|things | price |
|-------|-------|
|Laptop |  500  |
|Phone  |  800  | 
|Tablet |  300  |
|Watch  |  200  | 

Bar chart quickly shows:\
✅ Best-selling product\
✅ Worst-selling product\
✅ Easy comparison

That's why bar charts are everywhere:\
Business dashboards\
Sales reports\
University projects\
Power BI\
Data Science

## First Bar Chart

```python
import matplotlib.pyplot as plt

students = ["Ali", "Sara", "John", "Mary"]

marks = [85, 90, 78, 95]

plt.bar(students, marks)

plt.show()
```

## Add Title

```python
plt.title("Student Marks")
```

```python
import matplotlib.pyplot as plt

students = ["Ali", "Sara", "John", "Mary"]

marks = [85,90,78,95]

plt.bar(students, marks)

plt.title("Student Marks")

plt.show()
```

## Add Labels

```python
plt.xlabel("Students")

plt.ylabel("Marks")
```

```python
import matplotlib.pyplot as plt

students = ["Ali", "Sara", "John", "Mary"]

marks = [85,90,78,95]

plt.bar(students, marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

### Example

Monthly Sales

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr"]

sales = [1000,1500,2500,4000]

plt.bar(months, sales)

plt.title("Monthly Sales")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

Question: Which month sold most? \
Answer: April

Because highest bar = highest sales.

## Horizontal Bar Chart

Sometimes names are long.

plt.barh()

```python
import matplotlib.pyplot as plt

students = ["Ali","Sara","John","Mary"]

marks = [85,90,78,95]

plt.barh(students, marks)

plt.show()
```

The bars become horizontal.

## Add Grid

plt.grid()

```python
plt.bar(students, marks)

plt.grid()

plt.show()
```

Makes reading easier.

Compare Two Classes

Class A: class_a = [85,90,78,95]

Class B: class_b = [80,88,92,85]

For now just understand: Bar charts can compare groups.

## Mini Project 1

Student Marks Report

```python
import matplotlib.pyplot as plt

students = [
    "Ali",
    "Sara",
    "John",
    "Mary"
]

marks = [
    85,
    90,
    78,
    95
]

plt.bar(students, marks)

plt.title("Student Marks Report")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.grid()

plt.show()
```

## Mini Project 2

Cricket Runs

```python
import matplotlib.pyplot as plt

players = [
    "Babar",
    "Rizwan",
    "Virat",
    "Rohit"
]

runs = [
    120,
    95,
    110,
    80
]

plt.bar(players, runs)

plt.title("Player Runs")

plt.xlabel("Players")

plt.ylabel("Runs")

plt.show()
```

## Difference Between Line Chart & Bar Chart

### Line Chart 
**plt.plot()**

Best for: Change over time

Examples:
* Temperature
* Sales growth
* Stock prices

## Bar Chart
**plt.bar()**

Best for: Comparing categories

Examples:
* Student marks
* Product sales
* Employee performance

plot() = connect dots

bar() = standing buildings
