# Data Visualization Project

## What is a Data Visualization Project?
Imagine a school gives you marks:

|     |    |
|-----|----|
|Ali  | 85 |
|Sara | 90 |
|John | 78 |
|Mary | 95 |
|Tom  | 88 |

The principal asks:

"Can you show this data in different charts?"

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Mary", "Tom"],
    "Marks": [85, 90, 78, 95, 88]
}

df = pd.DataFrame(data)

print(df)
```

Who scored highest?

```python
import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Marks"])

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

What percentage of total marks belongs to each student?

```python
plt.pie(
    df["Marks"],
    labels=df["Name"],
    autopct="%1.1f%%"
)

plt.title("Marks Percentage")

plt.show()
```

How are marks distributed?

```python
plt.hist(
    df["Marks"],
    bins=5,
    edgecolor="black"
)

plt.title("Marks Distribution")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.show()
```

Find highest marks:

```python
print(df["Marks"].max())
```

Find lowest marks:

```python
print(df["Marks"].min())
```

Find average marks:

```python
print(df["Marks"].mean())
```

## Mini Challenge 2

Create Employee Salary Dataset

```python
data = {
    "Employee": ["Ali","Sara","John","Mary","Tom"],
    "Salary": [25000,30000,28000,35000,32000]
}
df = pd.DataFrame(data)

plt.bar(df["Employee"], df["Salary"])

plt.title("Employee Salary")

plt.xlabel("Employee")

plt.ylabel("Salary")

plt.show()
```

```python
plt.hist(
    df["Salary"],
    bins=5,
    edgecolor="black"
)

plt.title("Salary Distribution")

plt.xlabel("Salary")

plt.ylabel("Frequency")

plt.show()
```

```python
print(df["Salary"].max())
```

```python
print(df["Salary"].min())
```

```python
print(df["Salary"].mean())
```
