# Pie Charts (Showing Percentages & Parts of a Whole)

Imagine a Pizza 🍕 \
Suppose a pizza is divided among 4 friends:

Ali   = 40% \
Sara  = 30% \
John  = 20% \
Mary  = 10%

The whole pizza = 100%

A pie chart shows: Whole Circle = 100%

Each slice represents a percentage.

## What is a Pie Chart?
Simple definition: Pie Chart = **Shows parts of a whole**.

Examples:
* Market share
* Budget distribution
* Company departments
* Student grade distribution
* Mobile phone brands usage

A student's daily time:

Study      40% \
Sleep      30% \
Gaming     20% \
Exercise   10%

A pie chart immediately shows where most time is spent.

## Creating First Pie Chart

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]

plt.pie(sizes)

plt.show()
```

Python creates slices automatically.

## Add Labels

Without labels we don't know which slice belongs to whom.

```python
import matplotlib.pyplot as plt

labels = ["Ali", "Sara", "John", "Mary"]

sizes = [40,30,20,10]

plt.pie(
    sizes,
    labels=labels
)

plt.show()
```

### Student Example

```python
import matplotlib.pyplot as plt

subjects = [
    "Math",
    "Science",
    "English",
    "Computer"
]

marks = [
    90,
    85,
    75,
    100
]

plt.pie(
    marks,
    labels=subjects
)

plt.show()
```

## Show Percentages

Most important option:

```python
autopct="%1.1f%%"
```

```python
import matplotlib.pyplot as plt

labels = [
    "Study",
    "Sleep",
    "Gaming",
    "Exercise"
]

hours = [
    8,
    6,
    4,
    2
]

plt.pie(
    hours,
    labels=labels,
    autopct="%1.1f%%"
)

plt.show()
```

**What Does %1.1f%% Mean?**

Just remember: autopct="%1.1f%%"

means: Show percentage on chart.

## Add Title

```python
import matplotlib.pyplot as plt

activities = [
    "Study",
    "Sleep",
    "Gaming",
    "Exercise"
]

hours = [
    8,
    6,
    4,
    2
]

plt.pie(
    hours,
    labels=activities,
    autopct="%1.1f%%"
)

plt.title("Daily Activities")

plt.show()
```

This is exactly the type of chart we'll create with Matplotlib.

## Change Starting Position

```python
plt.pie(
    sizes,
    labels=labels,
    startangle=90
)
```

This rotates the pie. \
Think: Turn the pizza before cutting it.
    
## Explode a Slice

Want to highlight one slice?

```python
explode = [0.1, 0, 0, 0]
```

```python
plt.pie(
    sizes,
    labels=labels,
    explode=explode
)
```

Result: First slice comes slightly outside.

Useful for:\
Best-selling product\ 
Highest expense \
Top category
    
### Company Expenses

```python
import matplotlib.pyplot as plt

categories = [
    "Salary",
    "Rent",
    "Marketing",
    "Utilities"
]

expenses = [
    5000,
    2000,
    1500,
    500
]

plt.pie(
    expenses,
    labels=categories,
    autopct="%1.1f%%"
)

plt.title("Company Expenses")

plt.show()
```

Question: Where does most money go? \
Answer: Salary \
Largest slice.

**When Should We Use Pie Charts?**

Use Pie Charts when:
* Total = 100%
* Showing percentages
* Showing shares

Examples: \
Budget allocation \
Market share \
Department expenses \
Time usage

**When NOT To Use Pie Charts?**

Bad example: \
January \
February \
March \
April \
May \
June

This changes over time.

Use: **plt.plot()**

instead.

## Line vs Bar vs Pie

### Line Chart plt.plot()

Used for: Change over time \
Example: \
Sales growth \
Temperature \
Stock prices

### Bar Chart plt.bar()

Used for: Comparing categories \
Example: \
Student marks \
Product sales

### Pie Chart plt.pie()

Used for: \
Percentages \
Shares \
Parts of a whole

## Mini Project

Student Time Management

```python
import matplotlib.pyplot as plt

activities = [
    "Study",
    "Sleep",
    "Gaming",
    "Exercise"
]

hours = [
    8,
    6,
    4,
    2
]

plt.pie(
    hours,
    labels=activities,
    autopct="%1.1f%%"
)

plt.title("Student Daily Activities")

plt.show()
```

plt.pie() means Show percentages

labels= means Names of slices

autopct="%1.1f%%" means Show percentages

explode= means Pull slice outward
