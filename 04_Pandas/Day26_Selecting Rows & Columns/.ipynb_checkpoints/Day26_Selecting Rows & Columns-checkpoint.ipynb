# Selecting Rows and Columns
why?\
Because before analyzing data, we must know how to get the data we want. \
Think of a DataFrame as an Excel sheet.

| Name  | Age | Marks |
| ----- | --- | ----- |
| Ali   | 20  | 85    |
| Sara  | 21  | 90    |
| John  | 19  | 78    |
| David | 22  | 88    |

Sometimes we want:

Only Names \
Only Marks \
First row \
Last row \
Specific student

## Step 1: Create Dataset

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print(df)
```

### Column Selection

#### Select One Column

```python
print(df["Name"])
```

Think: DataFrame → Column → Series
    
#### Select Age Column

```python
print(df["Age"])
```

#### Select Multiple Columns
Use double brackets.

```python
print(df[["Name", "Marks"]])
```

### Easy Memory Trick

Single column  → ["Name"]

Multiple columns → [["Name","Marks"]]

Why double brackets?

Outer [] = List

Inner [] = Column Names

## Row Selection

Rows have indexes.

0 Ali \
1 Sara \
2 John \
3 David

#### Select Row Using loc
First Row


```python
print(df.loc[0])
```

### Third Row

```python
print(df.loc[2])
```

#### What is loc?

Think:

loc = Locate by Label

Use index numbers or row labels.

### Select Multiple Rows

```python
print(df.loc[[0, 2]])
```

#### iloc (Most Important)

Data Scientists use this constantly.

Think:

iloc = Integer Location

Select using position.

### First Row

```python
print(df.iloc[0])
```

### Second Row

```python
print(df.iloc[1])
```

## Difference Between loc and iloc

For now:

loc  → Label Based

iloc → Position Based

With normal datasets both may look similar.

### Select Specific Cell

Want: Sara's Marks

| Name | Age | Marks |
| ---- | --- | ----- |
| Ali  | 20  | 85    |
| Sara | 21  | 90    |

Sara's marks = 90

Using loc:

```python
print(df.loc[1, "Marks"])
```

Meaning: Row 1 Column Marks

### Using iloc

```python
print(df.iloc[1, 2])
```

Meaning: \
Row Position 1 \
Column Position 2
    
### Row Slicing

Like Python slicing.

First 2 rows:

```python
print(df.iloc[0:2])
```

First 3 rows:

```python
print(df.iloc[:3])
```

Last 2 rows:

```python
print(df.iloc[-2:])
```

### Row + Column Selection Together

Get: First 2 rows Only Name and Marks

```python
print(df.loc[0:1, ["Name", "Marks"]])
```

### Select Entire Column Using Dot Notation

```python
print(df.Name)
```

Works but professionals prefer:

```python
df["Name"]
```

because it always works.

### View First Row

```python
print(df.head(1))
```

### View First 2 Rows

```python
print(df.head(2))
```

### View Last 2 Rows

```python
print(df.tail(2))
```

## Real Data Scientist Example

Dataset:

```python
employees = {
    "Name": ["A", "B", "C", "D"],
    "Salary": [30000, 50000, 45000, 60000],
    "Department": ["IT", "HR", "IT", "Finance"]
}

df = pd.DataFrame(employees)
```

```python
#Get only salaries:

df["Salary"]
```

```python
#Get first employee:

df.loc[0]
```

```python
#Get employee C:

df.loc[2]
```

```python
#Get Salary column only:

df[["Salary"]]
```

These are everyday operations.

## Mini Project — Student Record Viewer

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(students)

print("Full Dataset")
print(df)

print("\nNames")
print(df["Name"])

print("\nMarks")
print(df["Marks"])

print("\nFirst Student")
print(df.loc[0])

print("\nThird Student")
print(df.loc[2])

print("\nSara's Marks")
print(df.loc[1, "Marks"])
```

### Memory Tricks 🧠

df["Name"] \
↓ \
One Column \
df[["Name","Marks"]] \
↓ \
Multiple Columns \
loc \
↓ \
Label\
iloc \
↓ \
Position \
loc[row, column] \
iloc[row_position, column_position]
