# Pandas Final Project

**Imagine You Work At A College**

The principal gives you this data: \
Student Name \
Age \
Marks 

Some students have missing data.

Your job:
1. Load Data
2. Check Missing Values
3. Clean Data
4. Create New Columns
5. Filter Students
6. Find Insights
7. Create Final Report

```python
import pandas as pd
import numpy as np

students = {
    "Name": ["Ali", "Sara", "John", "Mary", "Tom"],
    "Age": [20, np.nan, 19, 21, np.nan],
    "Marks": [85, 90, np.nan, 78, 95]
}

df = pd.DataFrame(students)

print(df)
```

```python
#Inspect Dataset
print(df.info())
```

```python
#Find Missing Values
print(df.isnull().sum())
```

```python
#Fill Age with average age.

df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
#Fill Marks with average marks.

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
```

```python
print(df.isnull().sum())
```

```python
# Principal gives everyone: +5 Bonus Marks
df["Bonus"] = 5
```

```python
#final marks:

df["Final_Marks"] = df["Marks"] + df["Bonus"]
```

```python
# Pass/Fail
df["Status"] = [
    "Pass" if x >= 80 else "Fail"
    for x in df["Final_Marks"]
]
print(df["Status"])
```

```python
# Show students scoring above 90.

top_students = df[df["Final_Marks"] > 90]

print(top_students)
```

```python
#Average Marks
print(df["Final_Marks"].mean())

#Highest Marks
print(df["Final_Marks"].max())

#Lowest Marks
print(df["Final_Marks"].min())

#Total Students
print(df["Name"].count())
```

```python
#Highest scorer first.

df = df.sort_values(
    by="Final_Marks",
    ascending=False
)

print(df)
```

```python
#Final Report
print("Average Marks:", df["Final_Marks"].mean())
print("Highest Marks:", df["Final_Marks"].max())
print("Lowest Marks:", df["Final_Marks"].min())
print("Total Students:", df["Name"].count())
```
