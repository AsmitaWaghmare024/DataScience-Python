# Pandas Filtering Data

In real jobs, datasets can contain:

10,000 rows \
100,000 rows \
1,000,000+ rows

We don't want all data.

We want only:

Students with marks above 80 \
Employees with salary above 50,000 \
Customers from London \
Products costing less than £100

This process is called **Filtering**

Think: Filtering = Keeping only the rows you want

### Create Dataset

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John", "David", "Emma"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 90, 78, 65, 95]
}

df = pd.DataFrame(students)

print(df)
```

## Greater Than (>)

Find students with marks above 80.

```python
print(df["Marks"] > 80)
```

### What Does True Mean?

True  = Keep this row

False = Ignore this row

### Actual Filtering

```python
print(df[df["Marks"] > 80])
```

### Memory Trick

df[condition]

Think: DataFrame[Rule]

Example:

```python
df[df["Marks"] > 80]
```

means: Show only rows where Marks > 80

## Less Than (<)

Students with marks below 80.

```python
print(df[df["Marks"] < 80])
```

## Equal To (==)

Students aged 20.

```python
print(df[df["Age"] == 20])
```

## Not Equal (!=)

Students not aged 20.

```python
print(df[df["Age"] != 20])
```

## Greater Than or Equal (>=)

```python
print(df[df["Marks"] >= 90])
```

## Less Than or Equal (<=)

```python
print(df[df["Marks"] <= 70])
```

## Multiple Conditions

Sometimes one condition isn't enough.

Example:

Marks > 80

AND

Age = 20

### AND Operator (&)

```python
print(df[(df["Marks"] > 80) & (df["Age"] == 20)])
```

**&** means **AND** \
Both conditions must be true.

### OR Operator (|)

Example: \
Age = 19

OR

Marks > 90

```python
print(df[(df["Age"] == 19) | (df["Marks"] > 90)])
```

**|** means **OR** \
At least one condition must be true.

## Filter Specific Columns After Filtering

Students scoring above 80.

Show only Names.

```python
print(df[df["Marks"] > 80]["Name"])
```

### Using loc with Filtering

Professional approach:

```python
print(df.loc[df["Marks"] > 80, ["Name", "Marks"]])
```

### Find Top Students

```python
top_students = df[df["Marks"] >= 90]

print(top_students)
```

### Find Maximum Marks Student

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John", "David", "Emma"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 90, 78, 65, 95]
}

df = pd.DataFrame(students)

highest = df["Marks"].max()

print(highest)
```

Get row:

```python
print(df[df["Marks"] == highest])
```

## Mini Project Student Performance Analyzer

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John", "David", "Emma"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 90, 78, 65, 95]
}

df = pd.DataFrame(students)

print("Full Dataset")
print(df)

print("\nStudents Above 80")
print(df[df["Marks"] > 80])

print("\nStudents Age 20")
print(df[df["Age"] == 20])

print("\nTop Students")
print(df[df["Marks"] >= 90])

print("\nAge 20 and Marks > 80")
print(df[(df["Age"] == 20) & (df["Marks"] > 80)])
```
