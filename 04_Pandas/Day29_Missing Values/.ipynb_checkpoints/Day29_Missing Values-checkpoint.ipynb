# Missing Values (NaN) in Pandas

Imagine a college database:

Name   |  Age |     Marks
-------|------|---------
Ali    |   20 |     85
Sara   |   ?  |     90
John   |   19 |     ?
Mary   |   21 |     78

Some information is missing.

Pandas represents missing data as: **NaN** \
Meaning: Not a Number

Think of NaN as: "I don't know the value."

## Step 1: Create Dataset with Missing Values

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "John", "Mary"],
    "Age": [20, np.nan, 19, 21],
    "Marks": [85, 90, np.nan, 78]
}

df = pd.DataFrame(data)

print(df)
```

What is np.nan? \
np means NumPy.

**import numpy as np**

NaN is a special value used to indicate missing data.

**np.nan**

## Step 2: Detect Missing Values

Use: **df.isnull()** \
Example:

```python
print(df.isnull())
```

Meaning: \
True  = Missing value found \
False = Value exists

**isnull()** asks: "Is something missing?"

### Count Missing Values

Instead of seeing every row:

```python
df.isnull().sum()
```

Meaning: \
Age column → 1 missing value \
Marks column → 1 missing value

**Most Common Interview Question**

How many missing values are in the dataset? \
Answer: df.isnull().sum()

## Step 3: Remove Rows with Missing Values

Suppose we only want complete rows. \
Use: **df.dropna()**

```python
print(df.dropna())
```

Rows containing NaN are removed.

**Understanding**

Before: 

|     |      |       |
|-----|------|-------|
|Ali  |   20 |    85 | 
|Sara |   NaN|    90 |
|John |   19 |    NaN| 
|Mary |   21 |    78 |

After: 

|     |       |      |
|-----|-------|------|
|Ali  |   20  |   85 |
|Mary |   21  |   78 |

## Step 4: Remove Columns with Missing Values

Use:

```python
df.dropna(axis=1)
```

Because:\
Age has NaN \
Marks has NaN \
Both columns are removed.
    
Remember Axis \
axis=0 → Rows \
axis=1 → Columns

## Step 5: Fill Missing Values
Most Data Scientists don't delete data. \
They fill missing values.

### Fill with Zero

```python
df.fillna(0)
```

### Fill Age with 18

```python
df["Age"] = df["Age"].fillna(18)

print(df)
```

**Example**

Suppose: Age missing \
Instead of deleting the student: Use average age

## Step 6: Fill with Mean

Dataset: \
20 \
NaN \
19 \
21 

Average: (20 + 19 + 21)/3 = 20

Fill missing value with 20.

Code:

```python
df["Age"] = df["Age"].fillna(
    df["Age"].mean()
)

print(df["Age"])
```

**Why Mean?**

Because it keeps data realistic. \
Better than 0 for age.

Nobody is 0 years old in a university dataset.

### Other Useful Statistics

Mean: **df["Age"].mean()**

Median: **df["Age"].median()**

Mode: **df["Age"].mode()[0]**

## Step 7: Check Again

After filling:

```python
df.isnull().sum()
```

```python
#Fix Marks Too

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
```

```python
df.isnull().sum()
```

No missing values.

**Workflow**

Whenever you get a CSV:\
1> df.head() 

Look at data. \
2> df.info()

Understand columns. \
3> df.isnull().sum()

Check missing values.\
4> fillna() or dropna()

Fix missing values.

## Mini Project: Student Database Cleaner

```python
import pandas as pd
import numpy as np

students = {
    "Name": ["Ali", "Sara", "John", "Mary"],
    "Age": [20, np.nan, 19, 21],
    "Marks": [85, 90, np.nan, 78]
}

df = pd.DataFrame(students)

print("Original Data")
print(df)

print("\nMissing Values")
print(df.isnull().sum())

df["Age"] = df["Age"].fillna(
    df["Age"].mean()
)

df["Marks"] = df["Marks"].fillna(
    df["Marks"].mean()
)

print("\nCleaned Data")
print(df)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

### Memory Tricks 🧠

Find Missing Values **df.isnull()** \
Think:Is anything missing?

Count Missing Values **df.isnull().sum()** \
Think: How many are missing?

Remove Missing Rows **df.dropna()** \
Think: Drop rows with missing data

Fill Missing Values **df.fillna(value)**\
Think: Replace missing data

Fill with Average \
**df["col"].fillna( \
    df["col"].mean() \
)**
