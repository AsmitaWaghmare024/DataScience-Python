# Pandas Adding & Removing Columns

Imagine This Table

| Name| Age  |  Marks|
|-----|------|-------|
|Ali  |  20  |    85 |
|Sara |  21  |    90 |
|John |  19  |    78 |

Teacher now wants: Pass / Fail column

New table:
    
|Name  |  Age | Marks | Status
|------|------|-------|--------
|Ali   |   20 |   85  |    Pass
|Sara  |   21 |   90  |    Pass
|John  |   19 |   78  |    Pass

We must add a new column.

### Create Dataset

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(students)

print(df)
```

## Adding a New Column

Syntax:
df["NewColumn"] = values

```python
df["City"] = ["London", "Manchester", "Birmingham"]

print(df)
```

**Memory Trick**

New Column

↓

df["ColumnName"] = data

### Create Column from Existing Columns

Very common in Data Science.

Example:

Teacher gives bonus marks.

```python
df["Final_Marks"] = df["Marks"] + 5

print(df)
```

Why Is This Useful?

Real companies create: Profit = Revenue - Cost

BMI = Weight / Height²

Final Marks = Marks + Bonus

### Create Pass/Fail Column

Suppose:

Pass if Marks >= 80

```python
df["Status"] = df["Marks"] >= 80

print(df)
```

```python
df["Status"] = ["Pass", "Pass", "Fail"]

print(df)
```

### Updating Existing Values

Suppose Ali's marks become 95.

```python
#Using loc:

df.loc[0, "Marks"] = 95

print(df["Marks"])
```

### Update Multiple Values

```python
df = pd.DataFrame({
    "Name": ["Ali", "Sara", "John"],
    "Marks": [85, 90, 78]
})

df["Marks"] = df["Marks"] + 10

print(df)
```

### Rename Columns

Suppose: Marks -> Exam_Marks

```python
df.rename(columns={"Marks": "Exam_Marks"})
```

### Permanent Rename

```python
df = df.rename(columns={"Marks": "Exam_Marks"})

print(df)
```

rename(): Changes column names
    
## Remove a Column

Suppose: Age column not needed

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(students)

df = df.drop("Age", axis=1)

print(df)
```

Why axis=1?

Remember: \
axis=0 → Rows \
axis=1 → Columns

We are deleting a column.

Therefore: axis=1
    
### Remove Multiple Columns

```python
students = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}
df = pd.DataFrame(students)
df = df.drop(["Age", "Marks"], axis=1)
print(df)
```

## Mini Project — Student Result System

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(students)

print("Original Dataset")
print(df)

df["Bonus"] = 5

df["Final_Marks"] = df["Marks"] + df["Bonus"]

df["Status"] = ["Pass", "Pass", "Fail"]

print("\nUpdated Dataset")
print(df)

df = df.rename(
    columns={"Marks": "Exam_Marks"}
)

print("\nRenamed Column")
print(df)

df = df.drop("Bonus", axis=1)

print("\nFinal Dataset")
print(df)
```
