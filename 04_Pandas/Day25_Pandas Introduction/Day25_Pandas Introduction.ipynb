## Step 2: Create a Series

Series = Single Column

Example:

```python
import pandas as pd

marks = pd.Series([80, 90, 70, 85])

print(marks)
```

### What Happened?

Pandas automatically creates labels.

|Index | Value |
 ------|-------|
|0     |  80   |
|1     |  90   |
|2     |  70   |
|3     |  85   |

Think: \
Series = One Column of Data 

Real Example:

```python
names = pd.Series(["Ali", "Sara", "John"])

print(names)
```

### Access Element

```python
print(names[0])
```

### Create DataFrame
Remember:

Series = One Column

DataFrame = Full Table

Create table:

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df)
```

### Understand DataFrame

Rows = Students\
Columns = Name, Age, Marks

        Columns
     ↓

Name | Age| Marks|
-----| ---| -----|
Ali  |  20|   85 |
Sara |  21|   90 |
John |  19|   78 |

↑
Rows

## Mini Project — Student Database

```python
import pandas as pd

students = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Age": [20,21,19,22],
    "Marks": [85,90,78,88]
}

df = pd.DataFrame(students)

print(df)

print("\nShape:")
print(df.shape)

print("\nAverage Marks:")
print(df["Marks"].mean())

print("\nHighest Marks:")
print(df["Marks"].max())

print("\nLowest Marks:")
print(df["Marks"].min())
```

## Memory Tricks 🧠

Series = One Column

DataFrame = Complete Table

head() = Top Rows

tail() = Bottom Rows

shape = (rows, columns)

columns = Column Names

info() = Dataset Information
