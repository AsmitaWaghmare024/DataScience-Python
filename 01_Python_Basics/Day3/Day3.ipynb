# DAY 3 — DECISION MAKING (if, else)

how to make decisions.

for example

        Imagine:

                If raining:
                       take umbrella
    
                Else: 
                       wear sunglasses

        That itself is programming logic.

```python
age = 25

if age >= 18:                   #if Means 'Check a condition' and 'Is age greater than or equal to 18?' 
                                #: Colon means:'Start decision block.'
        
    print("You can vote")       # This runs ONLY if condition is TRUE.
```

Suppose:

        age = 20

Then:

        20 >= 18
        TRUE

So computer prints: You can vote

WHAT IF CONDITION IS FALSE?

```python
age = 15

if age >= 18:
    print("You can vote")
```

Nothing happens, because condition is FALSE.

```python
age = 15

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

else means 'If condition is FALSE, do this instead'.
