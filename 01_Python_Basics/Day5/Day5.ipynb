# LISTS

This is HUGE.

Imagine storing 5 fruits.

Without lists:

```python
fruit1 = "apple"
fruit2 = "banana"
fruit3 = "mango"
fruit4 = "orange"
fruit5 = "grapes"
```

Messy and hard to manage.

LISTS SOLVE THIS

```python
fruits = ["apple", "banana", "mango"]
```

ONE variable stores MANY values.

That is a LIST.

Lists use [  ] square brackets.

```python
# ACCESSING ITEMS

fruits = ["apple", "banana", "mango"]

print(fruits[0])
```

WHY [0] ??

Because Python counting starts from 0.

| Position | Value  |
| -------- | ------ |
| 0        | apple  |
| 1        | banana |
| 2        | mango  |

This is called 'INDEX'

```python
numbers = [10, 20, 30, 40]

print(numbers[2])
```

CHANGING VALUES

```python
fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

#ADDING ITEMS

Using append()

```python
fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

#REMOVE ITEMS

```python
fruits.remove("banana")
```

#LIST LENGTH

Using len()

```python
fruits = ["apple", "banana", "mango"]

print(len(fruits))
```

LOOP THROUGH LIST

```python
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

for fruit in fruits Means take each item one by one.

CALCULATE TOTAL

```python
marks = [85, 72, 91, 67, 90]

total = sum(marks)

print(total)
```

CALCULATE AVERAGE

```python
average = sum(marks) / len(marks)  

print(average)
```

Student Marks Analyzer

```python
marks = []

for i in range(5):
    mark = int(input("Enter mark: "))
    marks.append(mark)

print("Marks:", marks)

print("Total:", sum(marks))

print("Average:", sum(marks) / len(marks))

print("Highest:", max(marks))

print("Lowest:", min(marks))
```

FUNCTIONS

| Function | Meaning     |
| -------- | ----------- |
| append() | add item    |
| remove() | delete item |
| len()    | count items |
| sum()    | total       |
| max()    | highest     |
| min()    | lowest      |


Grocery List

```python
grocery = ["milk", "bread", "eggs", "rice", "sugar"]

print("My Grocery List:")
for item in grocery:
    print("-", item)
```

Loop Through Lists

```python
numbers = [10, 20, 30, 40, 50]

for n in numbers:
    print("Number:", n, "Square:", n**2)
```

Shopping Cart

```python
cart = ["milk", "bread", "eggs", "rice"]

print("Items in your shopping cart:")
for item in cart:
    print("-", item)

cart.append("sugar")                      # Add a new item
print("\nUpdated cart:", cart)

cart.remove("bread")                     # Remove an item
print("After removing bread:", cart)
```

Favorite Movie List

```python
movies = ["Inception", "Interstellar", "The Dark Knight", "Avengers"]

print("My Favorite Movies:")
for m in movies:
    print("*", m)

if "Inception" in movies:                                # Check if a movie is in the list
    print("\nYes, Inception is one of my favorites!")
```

Highest Score Finder

```python
scores = [78, 92, 85, 67, 99, 74]

highest = max(scores)
print("Highest score is:", highest)

students = ["Asmita", "Ravi", "Neha", "Amit", "Priya", "Karan"]

topper_index = scores.index(highest)                                  # Find who got the highest (with index)
print("Topper is:", students[topper_index])
```
