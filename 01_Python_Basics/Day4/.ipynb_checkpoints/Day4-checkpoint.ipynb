#Patterns

```python
for i in range(1, 6):
    print(i * "*")
```

Loops teach computers "Do repetitive work automatically."

```python
for i in range(5, 0, -1):
    print(i * "*")
```

```python
for i in range(1, 6):
    print(i * str(i))
```

str(i) Means convert number into text.

Because " 3 " * 3 works.

But 3 * 3 gives 9

```python
#INCREASING NUMBERS

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")         #end=" " Normally print goes to new line. But end=" " keeps printing on SAME line.
    
    print()
```

```python
for i in range (1, 6):
    for j in range(1, i+1):
        print(j, end=" ")
        
    print()
```

here we used Nested Loop. This means loop inside another loop.

    Outer loop: controls rows

    Inner loop: controls columns/items

```python
#SQUARE PATTERN

for i in range(6):
    print("* " * 6)
```

```python
#RIGHT-ALIGNED TRIANGLE

for i in range(1, 6):
    print(" " * (5 - i) + "*" * i)   #" " * (5 - i) adds spaces. So stars shift right.
```

```python
#PYRAMID

for i in range(1, 6):
    print(" " * (5 - i) + "* " * i)  
```

```python
#INVERTED PYRAMID

for i in range(5, 0, -1):
    print(" " * (5 - i) + "* " * i)
```

```python
#FULL DIAMOND

# upper part

for i in range(1, 6):
    print(" " * (5 - i) + "* " * i)

# lower part

for i in range(4, 0, -1):
    print(" " * (5 - i) + "* " * i)
```

```python
#Alphabet Triangle

for i in range(1, 6):
    print(i * chr(64 + i))
```

```python
#Hollow Square

for i in range(5):          # 5 rows
    for j in range(5):      # 5 columns
        if i == 0 or i == 4 or j == 0 or j == 4:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Number Pyramid (fixed rows = 5)

for i in range(1, 6):                      # rows 1 to 5
    
    print(" " * (6 - i), end="")          # Print leading spaces
    
    for j in range(1, i+1):               # Print numbers from 1 to i
        print(j, end=" ")

    print()                               # Move to next line
```

```python
for i in range(1, 6):  # rows 1 to 5
    print(" " * (6 - i), end="")             # Print leading spaces to center the pyramid
    
    for j in range(i):                       # Print the same number i, repeated i times
        print(i, end=" ")
    print()                                  # Move to next line
```

```python
# Hollow Pyramid

for i in range(1, 6):                                 # 5 rows
    
    for j in range(1, 11):                           # width = 2*rows
        
        if j == 6 - i or j == 5 + i or i == 5:
            
            print("*", end="")
            
        else:
            print(" ", end="")
            
    print()
```

```python
# Floyd's Triangle


for i in range(1, 6):          # rows 1 to 5
    for j in range(1, i+1):    # print numbers in each row
        print((i*(i-1))//2 + j, end=" ")
    print()
```

```python
#Pascal's Triangle (Numbers)

for i in range(5):  # 5 rows
    num = 1
    print(" " * (5 - i), end="")         # spaces for centering
    for j in range(i+1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)  # formula for next value
    print()
```

```python
# Hollow DIAMOND

# Upper half

for i in range(1, 6):     # rows 1 to 5
    
    for j in range(1, 11):         # width = 2*rows
        
        if j == 6 - i or j == 5 + i:
            
            print("*", end="")
            
        else:
            
            print(" ", end="")
    print()

# Lower half

for i in range(4, 0, -1):          # rows 4 down to 1
    
    for j in range(1, 11):
        
        if j == 6 - i or j == 5 + i:
            
            print("*", end="")
            
        else:
            print(" ", end="")
    print()
```
