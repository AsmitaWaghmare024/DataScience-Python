Loop Through Dictionary

```python
student = {
    "name": "James",
    "age": 24,
    "city": "London"
}

for key, value in student.items():
    print(key, ":", value)
```

```python
student = {}

student["name"] = input("Enter Name: ")
student["age"] = int(input("Enter Age: "))
student["city"] = input("Enter City: ")

print("\nStudent Profile")

for key, value in student.items():
    print(key, ":", value)
```

# SETS

A Set is Collection of unique values No duplicates allowed.

```python
numbers = {1, 2, 3, 4, 4, 4, 2}

print(numbers)
```

Duplicates automatically removed.

Why Are Sets Useful?

Imagine

```python
students = [
    "James",
    "Rahul",
    "John",
    "Rahul",
    "John"
]
```

Need unique names.

```python
unique_students = set(students)

print(unique_students)
```

Add Items

```python
fruits = {"apple", "banana"}

fruits.add("mango")

print(fruits)
```

Remove Items

```python
fruits.remove("banana")
```

Check Membership

```python
fruits = {"apple", "banana", "mango"}

print("Guava" in fruits)
```

```python
cities = [
    "Mumbai",
    "Pune",
    "Mumbai",
    "Delhi",
    "Pune"
]
```

```python
unique_cities = set(cities)

print(unique_cities)
```

This is a very common data-cleaning task.

Unique Words Finder

```python
sentence = input("Enter a sentence: ")

words = sentence.lower().split()

unique_words = set(words)

print("Unique words:")

for word in unique_words:
    print(word)
```

Difference

| Feature            | List       | Set           | Dictionary     |
| ------------------ | ---------- | ------------- | -------------- |
| Ordered            | ✅          | ❌             | ✅              |
| Duplicates Allowed | ✅          | ❌             | Keys ❌         |
| Indexing           | ✅          | ❌             | By Key         |
| Use Case           | Store data | Unique values | Key-Value data |


Contact book dictionary

```python
contacts = {
    "Mom": "9876543210",
    "Dad": "9876543211",
    "Friend": "9876543212"
}

name = input("Enter a contact name: ")           # Ask user for a name

if name in contacts:
    print(name, ":", contacts[name])             # Check if name exists in dictionary
else:
    print("Sorry,", name, "not found in contacts.")
```
