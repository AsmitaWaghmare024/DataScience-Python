#SIMPLE CALCULATOR

combine everything

```python
n1 = input("Enter first number: ")

n2 = input("Enter second number: ")

print(n1 + n2)
```

-- What happens? --

Output becomes 105 NOT 15.

Because input() gives TEXT by default.
Computer thinks "10+ "5"

Text joining is called 'concatenation'.

Like  "Hello" + "World"

becomes: HelloWorld

-- HOW TO FIX THIS? --

We convert text into number.

Using ' int() ' means convert into integer number.

```python
n1 = int(input("Enter first number: "))

n2 = int(input("Enter second number: "))

print(n1 + n2)
```

This int() is like 'Computer treat this as number'.

For example:
When we fill online forms age, salary and marks systems often convert text into numbers internally.

Build Simple Calculator

```python
n1 = int(input("Enter first number: "))

n2 = int(input("Enter second number: "))

print("Addition:", n1 + n2)

print("Subtraction:", n1 - n2)

print("Multiplication:", n1 * n2)
```

REMEMBER FOREVER

Age Calculator

```python
# Input: birth year and current year

birth_year = int(input("Enter your birth year: "))
current_year = int(input("Enter the current year: "))

# Calculate age
age = current_year - birth_year

print("Your age is:", age)
```

 Marks Addition Program

```python
# add marks of 3 subjects

m1 = int(input("Enter marks for Subject 1: "))
m2 = int(input("Enter marks for Subject 2: "))
m3 = int(input("Enter marks for Subject 3: "))

total = m1 + m2 + m3
avg= total / 3

print("Total Marks:", total)
print("Average Marks:", avg)
```

Grocery bill calculator

```python
# i = item & p = price & q = quantity

i1_p = float(input("Enter price of item 1: "))
i1_q = int(input("Enter quantity of item 1: "))

i2_p = float(input("Enter price of item 2: "))
i2_q = int(input("Enter quantity of item 2: "))

i3_p = float(input("Enter price of item 3: "))
i3_q = int(input("Enter quantity of item 3: "))

# calculate total for each item
total1 = i1_p * i1_q
total2 = i2_p * i2_q
total3 = i3_p * i3_q

# final bill
grand_total = total1 + total2 + total3

print("Total Bill Amount:", grand_total)
```
