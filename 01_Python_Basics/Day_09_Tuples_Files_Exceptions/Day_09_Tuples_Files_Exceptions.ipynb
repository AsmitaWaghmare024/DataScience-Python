# PART 1  TUPLES

Suppose we have a list:

```python
fruits = ["apple", "banana", "mango"]

#We can change it:

fruits[0] = "orange"

print(fruits)
```

Python allows changes.

But What If We DON'T Want Changes?

Example:

Days in a week,
Months of a year,
Earth's gravity value,
Country codes

These should not accidentally change.

For that, Python gives us "TUPLE"

```python
days = ("Monday", "Tuesday", "Wednesday")
```

Notice

Lists use: [ ]     

Tuples use: ()

Think Like a Toy Box

List:

Open toy box
Can add/remove toys

Tuple:

Locked toy box
Can only look inside
Cannot change toys

```python
days = ("Monday", "Tuesday", "Wednesday")

print(days[0])
```

Same indexing as lists.

What Happens If We Try To Change It?

```python
days[0] = "Friday"
```

Python says "TypeError" Because tuples are protected.

```python
colors = ("Red", "Green", "Blue")

for color in colors:
    print(color)
```

# PART 2  FILE HANDLING

Until now 

name = input()

When program closes

Data disappears

Problem.

Writing To File

```python
file = open("notes.txt", "w")    #open() means Open notebook
                                 #"w" Means Write mode, Python can write inside notebook.
    
file.write("Hello World")        #write() Actually writes text.

file.close()                    #close() Close notebook safely.
```

Reading File

Suppose file contains Hello World

```python
file = open("notes.txt", "r")

content = file.read()

print(content)

file.close()
```

What Does "r" Mean?
"r" = Read

Think

Open notebook and read it

Suppose notebook already contains 'Apple'

We want 'Apple, Banana'

Use

```python
file = open("notes.txt", "a")        #a = Append means Add at end. Don't delete old content

file.write("\nBanana")

file.close()
```

Example

Student names:

```python
file = open("students.txt", "a")

name = input("Enter name: ")

file.write(name + "\n")

file.close()
```

Now every student is saved.

Instead of:

file = open(...)
...
file.close()

Use:

```python
with open("notes.txt", "r") as file:
    content = file.read()

print(content)
```

Python automatically closes file.

Safer.

Professional programmers use this.

# PART 3  EXCEPTION HANDLING

The Problem

age = int(input("Enter age: "))

User enters: twenty

Python crashes.

```python
age = int(input("Enter age: "))
```

Error: ValueError

Program stops.

Bad user experience.

Solution is Exception Handling.

Think program wears safety helmet.

When error comes Program survives.

```python
#Example

try:                                  # try: means Let's try this code.
    age = int(input("Enter age: "))   
    print(age)

except:                               # except: means If something goes wrong,run this code instead.
    print("Please enter a number")
```

What Happens?

User enters Twenty

Output is Please enter a number

Program does NOT crash.

That's exactly exception handling.

Common Error Example

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    print(num1 / num2)

except:
    print("Something went wrong")
```

Specific Errors

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    print(num1 / num2)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Enter numbers only")
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    print(num1 / num2)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Enter numbers only")
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    print(num1 / num2)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Enter numbers only")
```

Why This Is Important

Imagine 1 million rows of data

Some rows are broken.

Without exception handling Entire program crashes.

With exception handling Skip bad data

Continue processing.

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    print("Answer:", num1 / num2)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter valid numbers")
```
