VOWEL COUNTER

```python
text = input("Enter text: ")

count = 0

for char in text.lower():

    if char in "aeiou":          #Means check if character is vowel.
        count += 1

print("Vowels:", count)
```

PALINDROME CHECKER

Palindrome means same forward and backward.

Examples: madam,
level

```python
word = input("Enter word: ")

if word == word[::-1]:                    # [::-1] reverse string.
    print("Palindrome")
    
else:
    print("Not Palindrome")
```

When we type "Mom", Python compares it exactly to "moM" (the reversed string). Since "Mom" != "moM", it prints Not Palindrome.

To Fix this we Normalize the case

Convert the word to all lowercase (or all uppercase) before checking

```python
word = input("Enter word: ")

if word.lower() == word.lower()[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

We can also strip spaces so phrases like "nurses run" work:

```python
word = input("Enter word: ")
cleaned = word.replace(" ", "").lower()

if cleaned == cleaned[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

We can handle punctuation too 

```python
text = input("Enter text: ")

cleaned = ""

for char in text.lower():
    if char.isalnum():
        cleaned += char

if cleaned == cleaned[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
