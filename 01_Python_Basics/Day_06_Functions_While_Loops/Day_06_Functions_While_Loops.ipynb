SMALL PREVIEW

```python
count = 1               #count = 1 Start counting from 1.

while count <= 5:      #Means repeat while count is less/equal 5.
    print(count)
    
    count += 1         #means count = count + 1
```

Without this loop never stops.

while loops can become "infinite loops"

```python
while True:
    
    
    print("Hello")
```

```python
count = 1

while count <= 5:
    print(count)
```

This NEVER stops.

So while requires careful thinking.

PASSWORD SYSTEM EXAMPLE

```python
password = ""

while password != "python123":
    password = input("Enter password: ")

print("Access Granted")
```

Loop keeps running
UNTIL correct password entered.

break

Used to stop loop immediately.

```python
while True:
    name = input("Enter name: ")

    if name == "exit":
        break                    #means "Exit loop immediately.

    print(name)
```

continue

Skips current loop step.

```python
for i in range(5):

    if i == 2:
        continue

    print(i)
```

2 skipped.

```python
#ATM SYSTEM

balance = 1000

while True:

    print("\n1. Check Balance")
    print("2. Withdraw")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        print("Balance:", balance)

    elif choice == "2":
        amount = int(input("Enter amount: "))

        if amount <= balance:
            balance -= amount
            print("Remaining Balance:", balance)
        else:
            print("Insufficient Balance")

    elif choice == "3":
        print("Goodbye")
        break

    else:
        print("Invalid choice")
```

A function is "a reusable block of code."

Like a machine.

You create it once.
Use it many times.

--For example: Think of a coffee machine.

You press button snd coffee comes out.

Internally machine does many steps.

Functions are similar.

```python
def greet():
    print("Hello")
```

def means "I am creating a function."

greet: This is function name.

() : Functions use parentheses.

```python
#Nothing happens yet.
#we only CREATED function. To RUN it:

greet()
```

# WHY FUNCTIONS ARE IMPORTANT

Without functions:
programs become huge and messy.

Functions help:
reuse code,
organize code,
simplify projects.

#FUNCTION WITH INPUT

Functions can receive data.

```python
def greet(name):
    print("Hello", name) # name inside function is called parameter.It receives value.

greet("James")           # greet("James") means name = "James" inside function.
```

```python
def square(number):
    print(number * number)

num = int(input("Enter a number: ")) # take input from user
square(num)
```

RETURN VALUES

```python
def add(a, b):
    return a + b

result = add(5, 3)

print(result)
```

DIFFERENCE

print() Shows result.

return Sends result back.

Huge concept for future programming.
