# Standard Deviation 

Variance measured: How spread out the data is. \
Example: [70, 80, 90] \
Variance = 100 

But there is a problem.

- Problem with Variance

Marks are: 70, 80, 90

Unit = Marks

But variance becomes: 100 \
Actually: Marks² (square marks)

That is strange.

Nobody says: "My exam score is 100 square marks."

So we need a more natural measure.

## Solution

Take square root of variance. 

**Standard Deviation = √Variance**

For example: Variance = 100 \
√100 = 10 \
Standard Deviation = 10

Much easier to understand.

Standard Deviation tells us: **On average, how far values are from the mean.**

Imagine a target.\
Center: Mean \
Arrows near center: Small Standard Deviation \
Arrows everywhere: Large Standard Deviation

### Using Python

```python
import statistics

marks = [70,80,90]

print(statistics.stdev(marks))
```

### Population Version

```python
import statistics

marks = [70,80,90]

print(statistics.pstdev(marks))
```

## Why Two Versions?

Same reason as variance.

### Sample Standard Deviation 

**statistics.stdev()** \
Uses: n - 1

### Population Standard Deviation

**statistics.pstdev()** \
Uses: n

Used when entire population is available.

## Example

```python
import statistics

data = [10,20,30,40,50]

print("Mean:", statistics.mean(data))
print("Standard Deviation:", statistics.stdev(data))
```

Interpretation: Most values are about 16 units away from average.
