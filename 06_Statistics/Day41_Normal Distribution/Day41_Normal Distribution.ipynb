### Definition

A Normal Distribution is a distribution where:
Most values are near the average (mean).
Fewer values are very small.
Fewer values are very large.

It looks like a bell 

**Example 1: Student Marks**

In a class of 100 students:

10 students → below 50 \
80 students → between 60 and 85 \
10 students → above 90

Most students are in the middle.
This is approximately a normal distribution.

**Example 2: Human Height**

Imagine 1000 people. \
140 cm \
150 cm \
160 cm \
170 cm \
180 cm \ 
190 cm \
200 cm

Most people are around: 165–175 cm
Very few people are: 140 cm or 200 cm

**Example 3: Exam Scores**

Usually: Very Easy Exam \
Everyone gets high marks: Not normal.

Usually:Very Hard Exam \
Everyone gets low marks: Not normal.

A balanced exam: Few fail \
Most average Few full marks : Normal Distribution

### Important Property

In a perfect normal distribution: **Mean = Median = Mode**

Example: \
Mean = 80 \
Median = 80 \
Mode = 80

All are equal.

Let's generate data that looks normal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values
data = np.random.normal(70, 10, 1000)

plt.hist(data, bins=30, edgecolor="black")

plt.title("Normal Distribution")

plt.xlabel("Values")

plt.ylabel("Frequency")

plt.show()
```

**What does this mean?**

np.random.normal(70, 10, 1000) \
70 → Mean (average) \
10 → Standard deviation (spread) \
1000 → Number of values

Python creates 1000 numbers that follow a normal distribution.

### User Input Version

```python
import numpy as np
import matplotlib.pyplot as plt

mean = float(input("Enter Mean: "))
std = float(input("Enter Standard Deviation: "))
size = int(input("How many values to generate? "))

data = np.random.normal(mean, std, size)

plt.hist(data, bins=30, edgecolor="black")

plt.title("Normal Distribution")

plt.show()
```
