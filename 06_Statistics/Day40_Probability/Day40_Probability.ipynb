# Probability

Imagine a Coin

![Coin Image](coin.jpeg)
When you flip a coin, there are only 2 possible outcomes:

* Heads
* Tails

What is the chance of getting Heads? \
Answer: 1 out of 2 = 0.5 = 50%

**Probability Formula** \

Probability = Favorable Outcomes ÷ Total Outcomes

Think: “What I want” ÷ “All possible results”

### Example 1: Dice 🎲

A dice has 6 sides.

What is the probability of getting 4? \
Favorable outcomes = 1 (only the number 4) \
Total outcomes = 6 \
1 / 6  ≈ 0.167 ≈ 16.7%

### Example 2: Even Number on a Dice

Even numbers are: 2, 4, 6 \
Favorable outcomes = 3 \
Total outcomes = 6 \

3 / 6 = 1 / 2 = 0.5 = 50%

### Important Rule

Probability is always between 0 and 1

| Probability | Meaning   |
|-------------|-----------|
|     0       | Impossible|
|    0.5      | 50% Chance|    
|     1       | Certain   |


Examples

|  Event                    | Probability |
|---------------------------|-------------|
| Sun rises tomorrow        |    ≈ 1      |
| Roll a 7 on a normal dice |      0      |
| Get Heads on a coin       |     0.5     |


### Python Examples

**Coin**
Output: 0.5

**Dice**

Imagine you have a bag with balls.

🔴 🔴 🔴 🔵 🔵

There are: \
3 Red balls \
2 Blue balls

You close your eyes and pick one ball.

Question: What is the chance of getting a Red ball?

This chance is called **Probability**.

## Python Example 

```python
favorable = 1
total = 2

probability = favorable / total

print("Probability:", probability)
print("Percentage:", probability * 100, "%")
```

### Using User Input

```python
favorable = int(input("Enter favorable outcomes: "))
total = int(input("Enter total outcomes: "))

probability = favorable / total

print("Probability =", probability)
print("Percentage =", probability * 100, "%")
```

## Red Balls

```python
red = 3
blue = 2

total = red + blue

probability = red / total

print("Probability of Red Ball =", probability)
```

```python
red = int(input("Enter number of red balls: "))
blue = int(input("Enter number of blue balls: "))

total = red + blue

probability = red / total

print("Probability of Red Ball =", probability)
```
