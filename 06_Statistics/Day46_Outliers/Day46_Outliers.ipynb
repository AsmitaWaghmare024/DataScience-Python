# Outliers

## What is an Outlier?
An Outlier is a value that is much smaller or much larger than the rest of the data.

Think of it as: The value that doesn't fit with the group.

For Example: 

Imagine your class has these marks: 70 72 75 78 80 82 85 \
All students scored around 70–85. \
Now imagine one student scored: 200 \
Can exam marks be 200? No!

This value is very different from the others. It is called an **Outlier**.

**Example 1 – Salary**

Suppose a company has these monthly salaries (£): 2500 2700 2600 2800 2550 2650 50000

Everyone earns around £2500–£2800. \
One person earns £50000. \
That value is an Outlier.

**Example 2 – Student Marks**

65 68 70 72 74 75 5

Everyone scored around 70. \
One student scored 5. \
This is also an Outlier.

**Example 3 – Temperature**

28 29 30 31 29 30 150

Can today's temperature be 150°C? Impossible. \
Most likely it's a data entry error.

## Why Are Outliers Important?

Imagine a class: 70 72 75 78 80 82 85 

Average (Mean) ≈ 77 \
Now add: 500 

The average becomes much higher. \
One wrong value changed the whole result. \
That's why Data Scientists always check for Outliers.

**Memory Trick**

Think of a football team.

Everyone is wearing: Blue Jersey 🔵

One player is wearing: Pink Jersey 🩷

That player stands out. \
That's an Outlier.

## How Do We Find Outliers?

There are many methods. \
The most common is:

### IQR Method (Interquartile Range)
This is used in: \
Data Science \
Machine Learning \
Kaggle \
Research 

* Step 1 – Arrange Data

10 20 30 40 50 60 70 80 90

* Step 2 – Find Quartiles

From Day 45:\
Q1 = 30 \
Q2 = 50 \
Q3 = 70

* Step 3- Calculate IQR

Formula \
**IQR = Q3 - Q1**

Example \
70 - 30 = 40 \
So, IQR = 40

* Step 4 – Lower Limit

Formula **Q1 - 1.5 × IQR**

Example 30 - (1.5 × 40) \
        30 - 60 = -30
        
* Step 5 – Upper Limit

Formula **Q3 + 1.5 × IQR**

Example 70 + 60 = 130

* Final Rule

Any value **Below -30 or Above 130** is an Outlier. 

```python
import numpy as np

data = [10,20,30,40,50,60,70,80,90,200]

q1 = np.percentile(data,25)
q3 = np.percentile(data,75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("Q1 =", q1)
print("Q3 =", q3)
print("IQR =", iqr)
print("Lower Limit =", lower)
print("Upper Limit =", upper)

print()

print("Outliers:")

for value in data:
    if value < lower or value > upper:
        print(value)
```

### Understanding the Code

**Step 1** \
q1 = np.percentile(data,25) \
Finds the 25th percentile (Q1).

**Step 2** \
q3 = np.percentile(data,75) \
Finds the 75th percentile (Q3).

**Step 3** \
iqr = q3 - q1 \
Calculates the Interquartile Range.

**Step 4** \
lower = q1 - 1.5 * iqr \
upper = q3 + 1.5 * iqr \
Calculates the lower and upper boundaries.

**Step 5** \
for value in data: Checks every number one by one.

**Step 6** \
if value < lower or value > upper:

If the value is outside the limits, it's an Outlier.

### User Input Program

```python
import numpy as np

data = []

n = int(input("How many numbers? "))

for i in range(n):
    number = float(input(f"Enter number {i+1}: "))
    data.append(number)

q1 = np.percentile(data,25)
q3 = np.percentile(data,75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print()

print("Q1 =", q1)
print("Q3 =", q3)
print("IQR =", iqr)
print("Lower Limit =", lower)
print("Upper Limit =", upper)

print()

print("Outliers")

found = False

for value in data:
    if value < lower or value > upper:
        print(value)
        found = True

if not found:
    print("No Outliers Found")
```
