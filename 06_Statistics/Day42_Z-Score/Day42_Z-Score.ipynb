# Z-Score

## What is Z-Score?

Imagine your teacher says: Ali scored 95 marks. \
Is 95 good? \
You don't know.

Now imagine another teacher says: The class average is 94 marks. \
Now 95 doesn't seem very special.

Another class: Average = 50 Ali = 95 \
Now 95 is amazing.

So...

A number alone tells us very little. We must compare it with the average. That's exactly what Z-Score does.

**Simple Definition** \
A Z-Score tells us: How far a value is from the average (mean).

Not just far... It tells us how many Standard Deviations away it is.

Example: \
60 70 80 90 100

Mean:80 \
Ali scored:90 \
Difference: 90 - 80 = 10 \
Suppose Standard Deviation is: 10 \
Then, Z = 10 / 10 = 1 \
Ali is 1 Standard Deviation above average.

### Formula

Where, \
x = Value \
μ = Mean \
σ = Standard Deviation

**Remember Like This**

Imagine everyone is standing in a line. \
Average student: \
Ali stands one step ahead. \
His Z-score +1 

John stands two steps behind. \
His Z-score -2

Interpretation

| Z-Score | Meaning            |
| ------- | ------------------ |
| 0       | Exactly average    |
| +1      | Above average      |
| +2      | Much above average |
| -1      | Below average      |
| -2      | Much below average |

### Example 1 

Student marks

```python
import statistics

marks = [60,70,80,90,100]

mean = statistics.mean(marks)
std = statistics.stdev(marks)

student = 90

z = (student - mean) / std

print("Mean:", mean)
print("Standard Deviation:", std)
print("Z-Score:", round(z,2))
```

Meaning \
Ali scored 0.63 Standard Deviations above average.

### Example 2 (Using User Input)

```python
import statistics

marks = []

n = int(input("How many students? "))

for i in range(n):
    mark = int(input(f"Enter mark {i+1}: "))
    marks.append(mark)

student = int(input("Enter student's mark: "))

mean = statistics.mean(marks)
std = statistics.stdev(marks)

z = (student - mean) / std

print()

print("Mean =", round(mean,2))
print("Standard Deviation =", round(std,2))
print("Z-Score =", round(z,2))
```

## Mini Project

```python
import statistics

marks = [55,60,70,75,80,90,95]

mean = statistics.mean(marks)
std = statistics.stdev(marks)

print("Mean =", mean)
print("SD =", round(std,2))

for mark in marks:
    z = (mark - mean)/std
    print(mark, "->", round(z,2))
```

Look for: \
Positive Z-scores \
Negative Z-scores \
Value closest to 0
    
### Question 1

Marks 70 80 90 100 110

Find : \
Mean \
Standard Deviation \
Z-score for 100

```python
 marks = [70, 80, 90, 100, 110]

mean = statistics.mean(marks)
std = statistics.stdev(marks)

print("Mean =", mean)
print("SD =", round(std,2))

for mark in marks:
    if mark == 100:
        z = (mark - mean)/std
        print(f"Z-Score for 100 -> {round(z,2)}")
```

**Q: What is a Z-score?**

Answer: \
A Z-score measures how many standard deviations a value is above or below the mean. A positive Z-score means the value is above average, a negative Z-score means it is below average, and a Z-score of 0 means it is exactly equal to the mean.
