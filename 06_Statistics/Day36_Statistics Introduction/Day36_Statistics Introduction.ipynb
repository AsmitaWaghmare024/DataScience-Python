## Topic 1: Mean (Average)

Imagine marks: 80 90 70 60 100

Add them: 80 + 90 + 70 + 60 + 100 = 400 \
Count numbers: 5 \
Mean: 400 ÷ 5 = 80

**Formula**

Mean = Sum of Values ÷ Number of Values

### Python Example

```python
marks = [80, 90, 70, 60, 100]

mean = sum(marks) / len(marks)

print(mean)
```

Imagine 5 friends sharing chocolates.\
80 chocolates total & 5 friends

Each gets: 80 ÷ 5

Mean is simply equal sharing.

## Topic 4: Range

Range means: Highest - Lowest

Example: 100 - 60 = 40

Range tells us: How spread out the data is

### Python Example

```python
marks = [80, 90, 70, 60, 100]

range_value = max(marks) - min(marks)

print(range_value)
```

Class A: 78 79 80 81 82 \
Range: 82 - 78 = 4 \
Students are similar.

Class B: 20 40 60 80 100 \
Range: 100 - 20 = 80 \
Students are very different.

## Complete Statistics Program

```python
marks = [80, 90, 70, 60, 100]

print("Total:", sum(marks))

print("Count:", len(marks))

print("Average:", sum(marks)/len(marks))

print("Highest:", max(marks))

print("Lowest:", min(marks))

print("Range:", max(marks)-min(marks))
```
