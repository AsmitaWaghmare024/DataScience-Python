# Percentiles & Quartiles

## What is a Percentile?
A Percentile tells us what percentage of data is below a particular value. \
For Example:

Imagine your class has 100 students. After the exam, the teacher says: Ali is in the 90th percentile.

What does that mean? \
It does NOT mean Ali scored 90 marks.

It means: \
Ali scored better than 90% of the students. \
Only 10% of students scored higher than Ali.

Think of it like ranking people in a race.

Suppose these are exam marks: 40 50 60 70 80 90 100 \
Teacher says: 80 is the 75th percentile \
Meaning: 75% of students scored 80 or less. \
Only 25% scored higher.

Common Percentiles

| Percentile | Meaning                  |
| ---------- | ------------------------ |
| 25th       | 25% of values are below  |
| 50th       | Half of values are below |
| 75th       | 75% of values are below  |
| 90th       | 90% of values are below  |
| 100th      | Highest value            |

## What are Quartiles?
Quartiles divide the data into 4 equal parts.

Imagine a pizza 🍕 Cut it into 4 equal slices. \
Each slice is a Quartile.

### Q1 (First Quartile)
Q1 = 25th Percentile \
Means: 25% of values are below this point.

### Q2 (Second Quartile)
Q2 = 50th Percentile \
Also called: Median \
Half the data is below. \
Half the data is above.

### Q3 (Third Quartile)
Q3 = 75th Percentile \
Means: 75% of values are below this point.

**Memory Trick** \
Q1 = 25% \
Q2 = 50% (Median) \
Q3 = 75%

Remember: Quartiles are just special percentiles.

**Example** \
Marks: 40 50 60 70 80 90 100 

Approximate values: \
Q1 = 55 \
Q2 = 70 \
Q3 = 85

```python
import numpy as np

marks = [40,50,60,70,80,90,100]

q1 = np.percentile(marks,25)
q2 = np.percentile(marks,50)
q3 = np.percentile(marks,75)

print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)
```

### Finding Other Percentiles

```python
import numpy as np

marks = [40,50,60,70,80,90,100]

print("10th Percentile =", np.percentile(marks,10))

print("90th Percentile =", np.percentile(marks,90))
```

### User Input Program

```python
import numpy as np

marks = []

n = int(input("How many marks? "))

for i in range(n):
    mark = int(input(f"Enter mark {i+1}: "))
    marks.append(mark)

print()

print("Marks =", marks)

print()

print("Q1 =", np.percentile(marks,25))
print("Median (Q2) =", np.percentile(marks,50))
print("Q3 =", np.percentile(marks,75))
```
