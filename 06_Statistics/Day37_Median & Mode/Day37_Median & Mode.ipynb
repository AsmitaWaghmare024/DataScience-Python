# Median & Mode
Median = Middle Value \
Mode = Most Repeated Value

These are called Measures of Central Tendency. \
Simple meaning:They tell us where the center of the data is.

Imagine 5 Students' Marks 60 70 80 90 100

## Median (Middle Value)

Think of students standing in a line.

60 70 80 90 100

Find the middle person: 60 70 [80] 90 100

Median = 80

### Important Rule

Data must be sorted first.

Wrong: 50 10 40 20 30 \
Correct: 10 20 30 40 50

Now find middle.

### What If There Are Even Numbers?

Example: 10 20 30 40

No single middle number.

Two middle numbers: 10 [20 30] 40

Take average: (20 + 30) / 2 = 25

Median = 25

#### Python Example

```python
import statistics

numbers = [10,20,30,40]

print(statistics.median(numbers))
```

## Mode

Mode means: Most Frequent Value or Most Repeated Value

Example 10 20 20 30 40

Which appears most? 20 \
Mode = 20

Mode = Most Occurring \
Both start with: Mo

### Python Example

```python
import statistics

numbers = [10,20,20,30,40]

print(statistics.mode(numbers))
```

## Comparing Mean, Median, Mode

Example: 10 20 20 30 100

### Mean

10+20+20+30+100 = 180 \
180/5 =36 \
Mean = 36

### Median

10 20 [20] 30 100 \
Median = 20

### Mode

Most repeated: 20 \
Mode = 20

Notice: \
Mean = 36 \
Median = 20 \
Mode = 20 

Different answers! \
This happens often in real data.

```python
import statistics

marks = [80,90,70,60,100]

print("Mean:", statistics.mean(marks))

print("Median:", statistics.median(marks))

print("Mode:", statistics.mode(marks))
```

### Why did Python return 80?

In newer Python versions (3.8+), statistics.mode() returns the first value when there are multiple values tied for highest frequency.

Because 80 is the first value in the list.

### What If There Are Two Modes

```python
marks = [80,80,90,90,70]
statistics.multimode(marks)
```

Instead of mode(), use: statistics.multimode()

Example:

```python
import statistics

marks = [80,80,90,90,70]

print(statistics.multimode(marks))
```

This shows all modes.

```python
import statistics

marks = [80,90,80,60,100]

print("Mean:", statistics.mean(marks))

print("Median:", statistics.median(marks))

print("Mode:", statistics.mode(marks))
```
