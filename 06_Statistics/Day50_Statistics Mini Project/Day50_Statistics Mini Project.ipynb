# Final Statistics Project

Create a program that:
* Takes marks from the user.
* Calculates basic statistics.
* Detects outliers.
* Calculates quartiles.
* Displays a confidence interval.
* Makes a simple hypothesis testing decision.

```python
#Required Libraries
import statistics
import numpy as np

#Take Input
marks = []

n = int(input("How many students? "))

for i in range(n):
    mark = float(input(f"Enter mark {i+1}: "))
    marks.append(mark)

#Display the Data    
print("\nStudent Marks")
print(marks)
```

```python
#Mean
print("\nMean:", statistics.mean(marks))
```

```python
#Median
print("Median:", statistics.median(marks))
```

```python
#Mode
try:
    print("Mode:", statistics.mode(marks))
except statistics.StatisticsError:
    print("Mode: No unique mode")
```

We use try-except because some datasets may not have a unique mode.

```python
#Variance
print("Variance:", statistics.variance(marks))
```

```python
#Standard Deviation
print("Standard Deviation:", statistics.stdev(marks))
```

```python
#Minimum & Maximum
print("Minimum:", min(marks))
print("Maximum:", max(marks))
```

```python
#Quartiles

q1 = np.percentile(marks,25)
q2 = np.percentile(marks,50)
q3 = np.percentile(marks,75)

print("\nQuartiles")
print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
```

```python
#Percentiles
print("\n10th Percentile:", np.percentile(marks,10))
print("90th Percentile:", np.percentile(marks,90))
```

```python
# Outlier Detection (IQR Method)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("\nOutliers")

found = False

for mark in marks:
    if mark < lower or mark > upper:
        print(mark)
        found = True

if not found:
    print("No Outliers Found")
```

```python
#Confidence Interval 
#For simplicity, we'll use a fixed margin of error.

mean = statistics.mean(marks)

margin = 2

print("\n95% Confidence Interval")

print(mean-margin, "to", mean+margin)
```

```python
# Hypothesis Testing
p_value = float(input("\nEnter p-value: "))

if p_value < 0.05:
    print("Reject H0")
    print("Evidence supports H1")
else:
    print("Fail to Reject H0")
```

```python
#End Message....
print("\nStatistics Analysis Completed Successfully!")
```

```python
print("\nOutliers")

count = 0

for mark in marks:
    if mark < lower or mark > upper:
        print(mark)
        count += 1

if count == 0:
    print("No Outliers Found")

print("Total Outliers:", count)
```

```python
print("\nStudent Grades")

for mark in marks:

    if mark >= 90:
        grade = "A"

    elif mark >= 80:
        grade = "B"

    elif mark >= 70:
        grade = "C"

    elif mark >= 60:
        grade = "D"

    else:
        grade = "F"

    print(mark, "->", grade)
```
