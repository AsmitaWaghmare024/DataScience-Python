# Covariance (The Big Brother of Correlation)

Covariance → Tells Direction \
Correlation → Tells Direction + Strength

Suppose you observe students.

| Study Hours | Marks |
| ----------- | ----- |
| 1           | 40    |
| 2           | 50    |
| 3           | 60    |
| 4           | 70    |
| 5           | 80    |

As study hours increase, Marks also increase. \
They move together. \
Covariance helps measure this movement.

**Definition**

Covariance tells us: Do two variables move together or move in opposite directions? \
It only tells the direction.

### Example 1: Positive Covariance

| Temperature | Ice Cream Sales |
| ----------- | --------------- |
| 20          | 100             |
| 25          | 150             |
| 30          | 200             |
| 35          | 250             |

As temperature increases, Sales increase. \
↑ Temperature \
↑ Sales \
Positive Covariance ✅

### Example 2: Negative Covariance

| Speed | Travel Time |
| ----- | ----------- |
| 20    | 60          |
| 40    | 30          |
| 60    | 20          |
| 80    | 15          |

As speed increases, Time decreases. \
↑ Speed \
↓ Time \
Negative Covariance ✅

### Example 3: No Covariance

| Shoe Size | Marks |
| --------- | ----- |
| 6         | 80    |
| 9         | 60    |
| 7         | 90    |
| 10        | 70    |

No clear relationship. \
Covariance is near 0.

## Easy Memory Trick

Positive Covariance \
😊 😊 \
Both walk together

Negative Covariance \
😊 😢 \
One goes up \
One goes down

Zero Covariance \
🐱 🚗 \
No connection

## Covariance Formula

You don't need to memorize it yet, but know what it looks like.

Meaning: \
Take differences from averages \
Multiply them \
Find the average

Python will usually calculate it for us.

### Python Example

```python
import pandas as pd

data = {
    "Study_Hours":[1,2,3,4,5],
    "Marks":[40,50,60,70,80]
}

df = pd.DataFrame(data)

print(df.cov())
```

Focus on: 25.0 \
Positive value

Therefore: \
Positive Covariance
    
## Example 2: Negative Covariance

```python
import pandas as pd

data = {
    "Speed":[20,40,60,80],
    "Time":[60,30,20,15]
}

df = pd.DataFrame(data)

print(df.cov())
```

Meaning: Negative Covariance

### User Input Example

```python
import pandas as pd

x = []
y = []

n = int(input("How many values? "))

for i in range(n):
    a = float(input(f"Enter X value {i+1}: "))
    b = float(input(f"Enter Y value {i+1}: "))

    x.append(a)
    y.append(b)

df = pd.DataFrame({
    "X": x,
    "Y": y
})

print()

print(df)

print()

print("Covariance Matrix")

print(df.cov())
```

## Covariance Matrix

Just like Correlation Matrix.

```python
print(df.cov())
```

This table is called a **Covariance Matrix**

## Covariance vs Correlation

| Covariance                | Correlation                |
| ------------------------- | -------------------------- |
| Shows direction           | Shows direction + strength |
| Can be any number         | Always between -1 and 1    |
| Hard to compare           | Easy to compare            |
| Larger units affect value | Standardized               |

## Example

Covariance: 500 \
Good or bad? \
Hard to know. \
Correlation: 0.95 \

Immediately tells us: Strong Positive Relationship \
That's why Data Scientists usually prefer Correlation.

## Example

A company wants to know: Advertising Budget vs Sales

First: Calculate covariance. \
If positive: More advertising \
→ More sales \
Then calculate correlation to see how strong the relationship is.

## Mini Project

```python
import pandas as pd

data = {
    "Study":[2,3,4,5,6],
    "Marks":[50,60,70,80,90]
}

df = pd.DataFrame(data)

print(df)

print()

print("Covariance Matrix")

print(df.cov())
```

Observe: Is covariance positive or negative?

Q Can covariance be 0? \
Answer: Yes.

Q: What is Covariance? \
Answer: Covariance measures the direction of the relationship between two variables. A positive covariance means both variables tend to increase together, while a negative covariance means one increases as the other decreases.

## Remember

Covariance: Only tells Direction

Positive ↑ ↑

Negative ↑ ↓

Correlation \
Direction + Strength
