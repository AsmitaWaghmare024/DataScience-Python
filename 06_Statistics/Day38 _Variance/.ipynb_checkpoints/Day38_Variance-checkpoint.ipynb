# Variance

Imagine Two Classes

Class A
Marks: [78, 79, 80, 81, 82]

Class B
Marks: [20, 50, 80, 90, 160]

Both have nearly the same average. \
But are they similar? \
No!

Class A students scored almost the same. \
Class B students scored very differently.

We need a way to measure: "How spread out is the data?"

That measurement is called Variance.

**Example:**

Imagine 5 friends. \
Group 1 Heights: 170 171 169 170 170 \
Almost same height.

Group 2 Heights: 150 160 170 180 190 \
Big differences.

Group 2 has higher spread. \
Higher spread = Higher variance.

**Variance tells us:** How far numbers are from the average.

## Example

Marks: marks = [70, 80, 90]

**Step 1: Find Mean** \
Mean: \
(70 + 80 + 90) ÷ 3 \
= 240 ÷ 3 \
= 80

Mean = 80

**Step 2: Find Distance from Mean**  \
70 - 80 = -10 \
80 - 80 = 0 \
90 - 80 = 10

Table:

| Marks | Difference |
| ----- | ---------- |
| 70    | -10        |
| 80    | 0          |
| 90    | 10         |


Problem

If we add: -10 + 0 + 10 = 0 \
Everything cancels. \
Not useful.

**Step 3: Square Differences**

Square means: number × number \
Examples: 10² = 100 \
5² = 25 -> (-10)² = 100

Now:

| Difference | Squared |
| ---------- | ------- |
| -10        | 100     |
| 0          | 0       |
| 10         | 100     |

**Step 4: Average of Squared Differences**

(100 + 0 + 100) ÷ 3 \
200 ÷ 3 \
66.67

## Interpretation

#### Variance: 0
Means all values are identical. \
Example: [80,80,80,80] \
No spread. \
Perfectly equal.

#### Small variance:
[78,79,80,81,82]  Very close values.

#### Large variance: 
[20,50,80,120,180]  Very spread out.

Using Python

```python
import statistics

marks = [70,80,90]

print(statistics.variance(marks))
```

You may ask: "Teacher, we got 66.67 before!"

## Why Python Gives 100
Python's variance() calculates Sample Variance.

Formula: Divide by (n-1)

Instead of: Divide by n

For: [70,80,90] \
Squared differences: \
100 + 0 + 100 = 200 \
n = 3

Sample variance: 

200 ÷ (3-1) \
200 ÷ 2 \
100 

Output: 100

### Population Variance

If you want divide by n:

```python
import statistics

marks = [70,80,90]

print(statistics.pvariance(marks))
```

## Data Science Rule

Most of the time: **statistics.variance()** is used.

Remember:

variance()  -> sample variance

pvariance() -> population variance

#### Example 

```python
import statistics

data = [10,20,30,40,50]

print("Variance:", statistics.variance(data))
```

Large variance because numbers are spread out.

#### Example

```python
import statistics

data = [79,80,81,82,78]

print("Variance:", statistics.variance(data))
```

Small number. Because values are close together.

Imagine: Mean = Target \
Variance asks: "How far are the arrows from the target?" \
Closer arrows: Small Variance \
Farther arrows: Large Variance
