# Correlation

Imagine you have a plant 🌱. \
Every day you give it water.

| Water (Liters) | Plant Height (cm) |
| -------------: | ----------------: |
|              1 |                10 |
|              2 |                15 |
|              3 |                20 |
|              4 |                25 |
|              5 |                30 |

Question: As water increases, what happens to the plant? \
Answer: Plant grows taller. 

The two things move together. \
This is called **Correlation**.

### Definition
Correlation tells us: How two things are related.

It answers questions like: \
Does studying more increase marks? \
Does sleeping more increase energy? \
Does exercising reduce weight?

#### Example 1 – Positive Correlation

Suppose students study.

| Hours Studied | Marks |
| ------------: | ----: |
|             1 |    40 |
|             2 |    50 |
|             3 |    60 |
|             4 |    70 |
|             5 |    80 |

As study hours increase, Marks increase. \
This is called **Positive Correlation** 📈

Remember: ↑ One goes up, ↑ Other also goes up

### Example 2 – Negative Correlation

Imagine a car.

| Speed | Time Taken |
| ----: | ---------: |
|    20 |         60 |
|    40 |         30 |
|    60 |         20 |
|    80 |         15 |

As speed increases, Time decreases.\
This is **Negative Correlation 📉

Remember: ↑ One goes up, ↓ Other goes down

### Example 3 – No Correlation

| Shoe Size | Exam Marks |
| --------: | ---------: |
|         7 |         85 |
|         9 |         60 |
|         8 |         95 |
|        10 |         72 |

Does shoe size affect marks? No.
This is **No Correlation**

**Easy Memory Trick**

Positive \
😊😊 \
Both friends walk together.

Negative \
😊😢 \
One goes up \
One goes down.

No Correlation \
🐶🚗 \
No relationship.

## Correlation Coefficient
Scientists measure correlation using a number.

It is called r \
or \
Correlation Coefficient

Its value is always between -1 to +1

#### Meaning of Values

| Value | Meaning                      |
| ----- | ---------------------------- |
| +1    | Perfect Positive Correlation |
| +0.8  | Strong Positive              |
| +0.5  | Moderate Positive            |
| 0     | No Correlation               |
| -0.5  | Moderate Negative            |
| -0.8  | Strong Negative              |
| -1    | Perfect Negative             |

**Remember**

+1 Best Friends ❤️

0 Don't know each other 🙂

-1 Opposite Enemies 😄

## Python Example

```python
import pandas as pd

data = {
    "Study_Hours":[1,2,3,4,5],
    "Marks":[40,50,60,70,80]
}

df = pd.DataFrame(data)

print(df.corr())
```

Meaning: Perfect Positive Correlation.
    
## Example 2

```python
import pandas as pd

data = {
    "Speed":[20,40,60,80],
    "Time":[60,30,20,15]
}

df = pd.DataFrame(data)

print(df.corr())
```

Meaning: Strong Negative Correlation.
    
## User Input Example

```python
import pandas as pd

x = []
y = []

n = int(input("How many values? "))

for i in range(n):
    a = float(input(f"Enter X value {i+1}: "))
    b = float(input(f"Enter Y value {i+1}: "))

    x.append(a)
    y.append(b)

df = pd.DataFrame({
    "X": x,
    "Y": y
})

print()

print(df)

print()

print("Correlation")

print(df.corr())
```

Python calculates the correlation automatically.

### Correlation Matrix
Suppose you have

| Study | Sleep | Marks |
| ----: | ----: | ----: |
|     2 |     8 |    60 |
|     3 |     7 |    70 |
|     4 |     6 |    80 |

print(df.corr())

|         |Study| Sleep|Marks|
|---------|-----|------|-----|
|Study    | 1.00|-0.80 | 0.95|
|Sleep    |-0.80| 1.00 |-0.70|
|Marks    | 0.95| -0.70| 1.00|

This table is called a **Correlation Matrix**

### Example:

A company wants to know: Does advertising increase sales? \
They calculate the correlation. \
If 0.95 Advertising and sales are strongly related.

**Important Note**

Correlation does NOT mean Causation. \
Example: \
Ice cream sales increase. \
People drowning also increase. \
Does ice cream cause drowning? No \
The real reason is Summer \
People buy more ice cream and swim more. \
So: \
Correlation means things move together. \
It does not prove one causes the other.
