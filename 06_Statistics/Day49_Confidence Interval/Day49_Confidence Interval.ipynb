# Confidence Interval

Suppose you want to know: What is the average height of all students in your university? \
There are 20,000 students.

Will you measure every student? No. \
You measure only 200 students.

The average height of those 200 students is: 170 cm

But is the true average exactly 170 cm? \
Maybe it's: \
169 cm \
171 cm \
170.5 cm

We can't be 100% sure. \
So instead of saying: Average = 170 cm

We say: Average is probably between \
168 cm and 172 cm.

This range is called a **Confidence Interval**.

## Definition

A Confidence Interval (CI) is: A range of values that is likely to contain the true population value. \
Instead of giving one answer, we give a range.

### Example 1 – Student Marks

You ask 100 students. \
Average marks: 75 \
Confidence Interval: 73 to 77

Meaning: The true average marks are likely between 73 and 77.

### Example 2 – Salary

A survey finds: \
Average salary: £35,000 \
Confidence Interval: £34,000 to £36,000 \
The true average salary is likely inside this range.

## Why Not Give One Number?

Imagine you throw a dart. \
Will it always hit the exact center? No. \
It lands close to the center. \
Similarly, when we estimate from a sample, we are close, but not exact. \
That's why we use a range.

## Memory Trick

Think of weather forecasts. \
Instead of saying: Tomorrow's temperature = 30°C \
They often say: 29°C to 31°C \
That's a confidence interval.

## What is Confidence Level?

Suppose we say: 95% Confidence Interval

Does it mean we're 95% sure this one interval is correct? Not exactly.

It means: \
If we repeated the sampling process many times, about 95% of the intervals we calculate would contain the true population value.

In practice, you'll usually see: \
90% \
95% ✅ (most common) \
99%

## Margin of Error

Confidence Interval is built using: Estimate ± Margin of Error

### Example

Average: 75

Margin of Error: 2

Confidence Interval: 75 ± 2

Result: 73 to 77

### Remember Forever

Confidence Interval = Estimate ± Margin of Error

```python
mean = 75

margin = 2

lower = mean - margin

upper = mean + margin

print("Confidence Interval:")

print(lower, "to", upper)
```

### Understanding the Code

**Step 1** \
mean = 75 \
Our estimated average.

**Step 2** \
margin = 2 \
The possible error.

**Step 3** \
lower = mean - margin \
Lower limit.

**Step 4** \
upper = mean + margin \
Upper limit.

**Step 5** \
Print the interval.

## User Input Program

```python
mean = float(input("Enter sample mean: "))

margin = float(input("Enter margin of error: "))

lower = mean - margin

upper = mean + margin

print()

print("95% Confidence Interval")

print(lower, "to", upper)
```
