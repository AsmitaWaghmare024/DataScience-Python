# Hypothesis Testing

You create a new study method.

Usually students score: 70 marks \
After using your method, 30 students score: 75 marks \
Now the question is: Did your study method really improve marks, or was it just luck?

Hypothesis Testing helps answer this.

## What is a Hypothesis?

A hypothesis is simply: A statement that we want to test.

Example:

"This new teaching method improves marks." \
This is a **hypothesis**.

## Two Hypotheses
In statistics we always create TWO hypotheses.

###  1. Null Hypothesis (H₀)

Pronounced: H Zero \
This means: Nothing special happened. or No difference exists.

Example: \
New teaching method does NOT improve marks. This is H₀.

### 2. Alternative Hypothesis (H₁)

This means: Something changed. or There IS a difference.

Example: \
New teaching method improves marks. This is H₁.

### Example 1 – Medicine

A company creates a new medicine. \
H₀: The medicine does not work. \
H₁: The medicine does work.

### Example 2 – Website

A company changes the color of a button. \
H₀: Green button gets the same clicks as blue button. \
H₁: Green button gets more clicks.

## The p-value (Most Important Part)

Think of the p-value as: “How surprising is this result if H₀ is true?”

**Example**

Suppose H₀ says: “The study method does nothing.” \
But students scored much higher. \
We calculate a p-value.

Small p-value \
Very unlikely by chance \
↓ \
Evidence is strong. 

Large p-value \
Could easily happen by chance \
↓ \
Evidence is weak.

### The Famous Rule

Most people use: 0.05

This is called the significance level.

Symbol: α = 0.05

### Decision Rule
If p < 0.05

Reject H₀ \
Accept H₁

If p ≥ 0.05

Fail to reject H₀ \
Keep H₀

## Remember 

Small p-value \ 
↓ \
Strong evidence \
↓ \
Reject H₀

Large p-value \
↓ \
Weak evidence \
↓ \
Keep H₀

### Example 1

Suppose: p = 0.02 \
Compare with: 0.05 \
Since: 0.02 < 0.05 \
Decision: Reject H₀

The new method probably works.

### Example 2

Suppose: p = 0.40 \
Since: 0.40 > 0.05 \
Decision: Keep H₀

Not enough evidence.

## Simple Decision

```python
p_value = 0.03

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

H₀ \
No improvement. \
H₁ \
Improvement exists.

After testing: p = 0.01 \
Decision: Reject H₀

Netflix keeps the new system.

**Example**

H₀ \
Exercise does not reduce stress.

Q: What is Hypothesis Testing? \
Answer: Hypothesis Testing is a statistical method used to determine whether there is enough evidence in a sample of data to support a claim about a population. It involves comparing a Null Hypothesis (H₀) and an Alternative Hypothesis (H₁) using a p-value.
