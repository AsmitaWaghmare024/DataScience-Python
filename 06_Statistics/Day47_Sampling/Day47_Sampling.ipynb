# Sampling

Suppose there is a huge jar full of candies. \
There are 10,000 candies. 

You want to know: "Are these candies sweet?"
Will you eat all 10,000 candies? No! \
You'll eat only one or two candies. \
If they taste sweet, you'll guess that most candies are sweet. \
This is called **Sampling**.

## What is Population?

A Population means: The entire group that you want to study.

Examples: \
All students in your university \
All people in the UK \
All employees in a company \
Every customer of Amazon

Example \
Coventry University has: 20,000 Students \
All 20,000 students are called the **Population**.

## What is a Sample?

A Sample is: A small part of the population chosen for study.

Example: \
Population: 20,000 students

Sample: 200 students \
We study only 200 students.

## Why Don't We Study the Whole Population?

Imagine asking every person in the UK: "Which mobile phone do you use?"

There are millions of people. \
It would take: \
Too much time \
Too much money \
Too much effort 

Instead, survey 2,000 people. \
This gives a good estimate.

### Example 1 – Election

Before elections, TV channels ask: 2,000 people \
Not: 70 million people \
That 2,000 people is called a Sample. 

### Example 2 – Hospital

A new medicine is made. \ 
Doctors don't give it to everyone immediately. \
They first test it on: 100 people \
Those 100 people are the Sample.

### Example 3 – Company

A company has: 1,00,000 customers \
It asks:500 customers \
for feedback. \
Those 500 customers are the sample.

#### Memory Trick

### Types of Sampling

There are many methods.

1. Random Sampling

Everyone has an equal chance of being selected.

Example: \
Names: Ali Sara John Emma David \
Python randomly selects: Sara David 

No one is favored.

2. Systematic Sampling

Select every nth person.

Example: \
Every 5th student. \
5 10 15 20 \
Easy to perform.

3. Stratified Sampling

Suppose a school has: \
Boys = 600 \
Girls = 400 

Take samples from both groups. \
This keeps the sample balanced.

4. Cluster Sampling

Suppose there are 100 schools. \
Instead of selecting students from all schools, \
Choose: School 5 School 10 School 25

Study everyone in those schools.

Summary Table

| Sampling Type | Meaning                         |
| ------------- | ------------------------------- |
| Random        | Anyone can be selected          |
| Systematic    | Every nth person                |
| Stratified    | Divide into groups, then sample |
| Cluster       | Select entire groups            |

## Random Sampling

```python
import random

students = [
    "Ali",
    "Sara",
    "John",
    "Emma",
    "David",
    "Tom",
    "Lily"
]

sample = random.sample(students,3)

print("Population")

print(students)

print()

print("Sample")

print(sample)
```

Notice: \
The sample changes each time you run the program.
    
### Understanding the Code

Step 1 \
import random

Imports Python's random module.

Step 2 \
students = [...] \
Creates the population.

Step 3 \
random.sample(students,3) \
Means: Select 3 random students.

### User Input Program

```python
import random

students = []

n = int(input("How many students? "))

for i in range(n):
    name = input(f"Enter student {i+1}: ")
    students.append(name)

sample_size = int(input("How many students in the sample? "))

sample = random.sample(students, sample_size)

print()

print("Population")

print(students)

print()

print("Random Sample")

print(sample)
```
