# SQL LIMIT Clause

## What is LIMIT? 
LIMIT tells SQL: "Only return this many rows."

Example: \
"I only want to announce the top 3 students."

I doesn't call all 50 students. \
Only calls: \
Emma – 95 \
Ali – 90 \
Sara – 85

Showing only a few records is exactly what **LIMIT** does.

### Without LIMIT:

*SELECT * FROM students;*

Returns all students.

### With LIMIT:

*SELECT * FROM students*
*LIMIT 3;*

Returns only the first 3 students.

## SQL Syntax

*SELECT column_name \
FROM table_name \
LIMIT number;*

**Example Table**

| ID | Name  | Age | Marks |
| -- | ----- | --- | ----: |
| 1  | Ali   | 22  |    90 |
| 2  | Sara  | 21  |    85 |
| 3  | John  | 23  |    78 |
| 4  | Emma  | 20  |    95 |
| 5  | David | 22  |    60 |

### Example 1 – First 3 Students

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
LIMIT 3
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

### Example 2 – First 2 Student Names

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT name
FROM students
LIMIT 2
""")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

## Why LIMIT Alone Can Be Misleading
 
| Name  | Marks |
| ----- | ----: |
| Ali   |    90 |
| Sara  |    85 |
| John  |    78 |
| Emma  |    95 |
| David |    60 |

If write: \
*SELECT * \
FROM students \
LIMIT 3;*

Output: \
Ali \
Sara \
John \

These are not the top 3 students. \
They are simply the first 3 rows stored in the table.

## Correct Way: ORDER BY + LIMIT

I want the top students, first sort by marks.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks DESC
LIMIT 3
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## LIMIT + WHERE

Show only 2 students whose marks are greater than 80.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE marks > 80
LIMIT 2
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## ORDER BY + WHERE + LIMIT

This is one of the most common SQL queries.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE marks > 80
ORDER BY marks DESC
LIMIT 2;
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

### User Input Example

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

limit = int(input("How many students do you want to see? "))

query = f"""
SELECT *
FROM students
LIMIT {limit}
"""

cursor.execute(query)

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

Note: This is fine for learning. In real applications, avoid building SQL queries directly with f-strings for user input because of SQL injection risks.

## Mini Project 
Top 3 Students

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks DESC
LIMIT 3
""")

students = cursor.fetchall()

print("=" * 45)
print("        TOP 3 STUDENTS")
print("=" * 45)

rank = 1

for student in students:

    print(f"Rank  : {rank}")
    print(f"Name  : {student[1]}")
    print(f"Marks : {student[3]}")
    print("-" * 45)

    rank += 1

connection.close()
```

Q1: What is the purpose of LIMIT? \
Answer: LIMIT restricts the number of rows returned by a query.

Q2: Why is ORDER BY often used with LIMIT? \
Answer: Because LIMIT alone returns the first rows in storage order, while ORDER BY ensures the rows are sorted first (for example, highest marks first).

Q3: Which clause comes first, ORDER BY or LIMIT? \
Answer: \
*ORDER BY \
LIMIT* \
LIMIT always comes after ORDER BY.
