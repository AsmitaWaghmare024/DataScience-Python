## What is a Table?

A Table stores related information in rows and columns. \
Example: Students Table

| ID | Name | Age | Marks |
| -- | ---- | --- | ----- |
| 1  | Ali  | 22  | 90    |
| 2  | Sara | 21  | 85    |
| 3  | John | 23  | 78    |

### What is a Row?

A row stores one complete record. \
Example:

| ID | Name | Age | Marks |
| -- | ---- | --- | ----- |
| 2  | Sara | 21  | 85    |

This entire line is one **row**.

### What is a Column?

A column stores one type of information. \
Example:

## What is a Primary Key?

Suppose there are two students named Ali.

| ID | Name |
| -- | ---- |
| 1  | Ali  |
| 2  | Ali  |

How do we know which Ali is which? \
We use a **Primary Key**. \
The ID is unique. \
No duplicates. \
No empty values.

#### Rules of a Primary Key
* Must be unique 
* Cannot be empty 
* One primary key per table 

#### Memory Trick

Think of your passport number. \
Many people may have the same name. \
But nobody has the same passport number. \
A Primary Key works the same way.

## SQL Syntax – CREATE TABLE

```python
CREATE TABLE students (

    id INTEGER PRIMARY KEY,

    name TEXT,

    age INTEGER,

    marks REAL

);
```

#### Understanding the Code

CREATE TABLE students (Creates a table named students.)

id INTEGER PRIMARY KEY (Creates an ID column. Integer Unique Primary Key)

name TEXT (Stores student names.)

age INTEGER (Stores ages.)

marks REAL (Stores decimal marks.)

## Creating a Table Using Python

```python
import sqlite3

connection = sqlite3.connect("school.db")

cursor = connection.cursor()

cursor.execute("""
CREATE TABLE students (

    id INTEGER PRIMARY KEY,

    name TEXT,

    age INTEGER,

    marks REAL

)
""")

connection.commit()

print("Table Created Successfully!")

connection.close()
```

#### Understanding the Python Code

import sqlite3 (Import SQLite.)

connection = sqlite3.connect("school.db") (Open the database.)

cursor = connection.cursor() (The cursor is like a messenger.It sends SQL commands to the database.)

cursor.execute(...) (Executes the SQL command.)

connection.commit() (Saves the changes permanently. Without commit(), your table may not be saved.)

connection.close() (Closes the database. Always close it after use.)

### Common Error

If you run the program twice, you'll get: table students already exists \
Why? \
Because SQL doesn't allow creating the same table twice.

Fix \
Use:

CREATE TABLE IF NOT EXISTS students (

    id INTEGER PRIMARY KEY,

    name TEXT,

    age INTEGER,

    marks REAL

);

Now SQL checks first. \
If the table already exists, it won't create it again or show an error.

```python
import sqlite3

connection = sqlite3.connect("school.db")

cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (

    id INTEGER PRIMARY KEY,

    name TEXT,

    age INTEGER,

    marks REAL

)
""")

connection.commit()

print("Table Ready!")

connection.close()
```

## View All Tables

You can ask SQLite what tables exist.

```python
import sqlite3

connection = sqlite3.connect("school.db")

cursor = connection.cursor()

cursor.execute("""
SELECT name FROM sqlite_master
WHERE type='table';
""")

tables = cursor.fetchall()

print("Tables:")

for table in tables:
    print(table[0])

connection.close()
```

Create a table called employees.

Create three tables in a new database called college.db:

students \
id (INTEGER PRIMARY KEY) \
name (TEXT) \
age (INTEGER) \
marks (REAL) 

teachers \
teacher_id (INTEGER PRIMARY KEY) \
name (TEXT) \
subject (TEXT) \
salary (REAL) 

courses \
course_id (INTEGER PRIMARY KEY) \
course_name (TEXT) \
duration (TEXT)


```python
import sqlite3

connection = sqlite3.connect("college.db")

cursor = connection.cursor()

# Students Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")

# Teachers Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS teachers (
    teacher_id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT,
    salary REAL
)
""")

# Courses Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    duration TEXT
)
""")

connection.commit()

print("All Tables Created Successfully!")

connection.close()
```
