# Inserting Data into SQL Tables

Yesterday, created this table:

| ID | Name | Age | Marks |
| -- | ---- | --- | ----- |

But it's empty. \
Now Add Ali.

Now the table becomes:

| ID | Name | Age | Marks |
| -- | ---- | --- | ----- |
| 1  | Ali  | 22  | 90    |

This process is called **Inserting Data**.

## What is a Record?

A record is one complete row.

Example:  

| ID | Name | Age | Marks |
| -- | ---- | --- | ----- |
| 1  | Ali  | 22  | 90    |

This entire row is one record.

## SQL INSERT INTO

The SQL command is:

INSERT INTO students  \
VALUES (1,'Ali',22,90);

Meaning: Insert one row into the students table.

INSERT INTO students (Means:Put data into the students table.) \
VALUES(...) Contains the values to insert.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")

connection.commit()

print("Students table created!")

connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT name FROM sqlite_master
WHERE type='table';
""")

print(cursor.fetchall())

connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("college.db")  #Open the database.
cursor = connection.cursor()    #Create a cursor.

cursor.execute("""
INSERT INTO students
VALUES (1,'Ali',22,90)
""")                                 #Execute the SQL command.

connection.commit()     #Save the record permanently. Without commit(), the data may not be saved.

print("Record Inserted Successfully!")

connection.close()
```

```python
connection = sqlite3.connect("college.db")  #Open the database.
cursor = connection.cursor()    

cursor.execute("""
INSERT INTO students
VALUES (2,'Sara',21,85)
""")

cursor.execute("""
INSERT INTO students
VALUES (3,'John',23,78)
""")

connection.commit()
```

## Parameterized Queries

Never build SQL using string concatenation.

Instead, use placeholders (?).

```python
cursor.execute(
"""
INSERT INTO students
(id,name,age,marks)
VALUES (?,?,?,?)
""",
(5,"David",20,88)
)

connection.commit()
```

Why? \
Prevents SQL Injection. \
Automatically handles quotes. \
Professional coding style.

Think of ? as Empty Boxes \
ID      → ? \
Name    → ? \
Age     → ? \
Marks   → ? 

Python fills the boxes with: (5,"David",20,88)
    
## User Input Program

```python
import sqlite3

connection = sqlite3.connect("college.db")

cursor = connection.cursor()

student_id = int(input("Enter ID: "))
name = input("Enter Name: ")
age = int(input("Enter Age: "))
marks = float(input("Enter Marks: "))

cursor.execute("""
INSERT INTO students
(id,name,age,marks)
VALUES (?,?,?,?)
""",(student_id,name,age,marks))

connection.commit()

print("Student Added Successfully!")

connection.close()
```

## Mini Project

Write a program that lets the user add 3 students.

```python
import sqlite3

connection = sqlite3.connect("college.db")

cursor = connection.cursor()

for i in range(3):

    print(f"\nStudent {i+1}")

    student_id = int(input("ID: "))
    name = input("Name: ")
    age = int(input("Age: "))
    marks = float(input("Marks: "))

    cursor.execute("""
    INSERT INTO students
    (id,name,age,marks)
    VALUES (?,?,?,?)
    """,(student_id,name,age,marks))

connection.commit()

print("\n3 Students Added Successfully!")

connection.close()
```

Q1: What is the purpose of INSERT INTO? \
Answer: It is used to add new records (rows) into a database table.

Q2: Why should we use parameterized queries? \
Answer: Parameterized queries improve security by preventing SQL Injection and make code easier to maintain.

Q3: What happens if you don't call commit()? \
Answer:The inserted data may not be permanently saved to the database.

Q4: What happens if you insert a duplicate Primary Key? \
Answer: SQLite raises a UNIQUE constraint failed error because a Primary Key must be unique.
