# SQL UPDATE Statement

## What is UPDATE?
UPDATE changes existing data inside a table. \
Think: \
Old Data \
     │ \
 UPDATE \
     │ \
New Data

### SQL Syntax

*UPDATE table_name \
SET column = value  \
WHERE condition;*

Example:

| ID | Name | Marks |
| -- | ---- | ----: |
| 1  | Ali  |    90 |
| 2  | Sara |    85 |
| 3  | John |    78 |

After the exam, Ali's marks become 95.

Do I erase the whole notebook? NO \
I chnage only Ali's marks. \
That is exactly what **UPDATE** does.

## Important Rule

UPDATE \
  ↓ \
SET \
  ↓ \
WHERE

Without WHERE, every row will be updated!

**Example Table**

| ID | Name | Age | Marks |
| -- | ---- | --- | ----: |
| 1  | Ali  | 22  |    90 |
| 2  | Sara | 21  |    85 |
| 3  | John | 23  |    78 |

## Example 1 – Update One Student
Ali's marks become 95.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
UPDATE students
SET marks = 95
WHERE id = 1
""")

connection.commit()

print("Record Updated Successfully!")

connection.close()
```

### Verify the Update

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

### Why commit() is Necessary?

Without: connection.commit() \
the changes stay only in memory.

After closing the database: Changes disappear.

Always write: connection.commit() \
after UPDATE.

## Example 2 – Update a Name

John changes his name to Johnny.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
UPDATE students
SET name = 'Johnny'
WHERE id = 3
""")

connection.commit()

print("Name Updated!")

connection.close()
```

## Example 3 – Update Multiple Columns

Ali's:\
Age -> 23 \
Marks -> 98

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
UPDATE students
SET age = 23,
    marks = 98
WHERE id = 1
""")

connection.commit()

connection.close()
```

## Example 4 – Update Every Student
Increase everyone's marks by 5.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
UPDATE students
SET marks = marks + 5
""")

connection.commit()

connection.close()
```

Notice: \
I did not write numbers. \
I used marks = marks + 5

**Example** \
Never do this accidentally.

*UPDATE students \
SET marks = 100;*

Result:

| Name | Marks |
| ---- | ----: |
| Ali  |   100 |
| Sara |   100 |
| John |   100 |
| Emma |   100 |
| David|   100 |

Every student's marks become 100 because there is no WHERE clause.

UPDATE without WHERE = Update EVERY row

## User Input Example

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

student_id = int(input("Enter Student ID: "))
new_marks = float(input("Enter New Marks: "))

cursor.execute("""
UPDATE students
SET marks = ?
WHERE id = ?
""", (new_marks, student_id))

connection.commit()

print("Marks Updated Successfully!")

connection.close()
```

Verify

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()
cursor.execute("""
SELECT *
FROM students
""")

print(cursor.fetchall())
```

## Mini Project – Student Marks Updater

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

student_id = int(input("Enter Student ID: "))
new_marks = float(input("Enter New Marks: "))

cursor.execute("""
UPDATE students
SET marks = ?
WHERE id = ?
""", (new_marks, student_id))

connection.commit()

cursor.execute("""
SELECT *
FROM students
WHERE id = ?
""", (student_id,))

student = cursor.fetchone()

print("\nUpdated Record")
print("-"*30)
print("ID    :", student[0])
print("Name  :", student[1])
print("Age   :", student[2])
print("Marks :", student[3])

connection.close()
```

Q1: What is the purpose of UPDATE? \
Answer: UPDATE is used to modify existing records in a table.

Q2: Why is the WHERE clause important with UPDATE? \
Answer: Without WHERE, every row in the table is updated.

Q3: Why is commit() necessary after UPDATE? \
Answer: commit() permanently saves the changes to the database.

Q4: Can UPDATE modify multiple columns? \
Answer: Yes.

Example: 

*UPDATE students \
SET age = 23, \
    marks = 98  \
WHERE id = 1;*
