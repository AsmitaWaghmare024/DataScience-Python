## What is SELECT?

SELECT is the SQL command used to retrieve data from a table. \
Think of it like asking questions to the database.

Examples:
* Show all students.
* Show only names.
* Show only marks.

### SQL Syntax

SELECT column_name \
FROM table_name;

**What does * mean?**

*means: All Columns

*Instead of writing:*

SELECT id, name, age, marks
FROM students;

*we simply write*

SELECT *
FROM students;


```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS teachers (
    teacher_id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT,
    salary REAL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    duration TEXT
)
""")

connection.commit()

print("All tables created!")

connection.close()
```

### Understanding fetchall()

Think of a classroom. \
Teacher says: Bring every student's notebook. \
That is what fetchall() does.

**rows = cursor.fetchall()**

It returns all rows as a list.

### Why Are There Parentheses?

Each row becomes a tuple.

Example: (1,'Ali',22,90)

Many rows become

[
(1,'Ali',22,90),
(2,'Sara',21,85),
(3,'John',23,78)
]

### Display Data Properly

Instead of print(students) \
Use \
for student in students: \
    print(student)
   

```python
for student in students:

    print("ID    :", student[0])
    print("Name  :", student[1])
    print("Age   :", student[2])
    print("Marks :", student[3])

    print("-----------------------")
```

## Selecting One Column

Suppose I only need student names.

SELECT name FROM students;

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("SELECT name FROM students")

names = cursor.fetchall()

for name in names:
    print(name)

connection.close()   # Close only after all work is done
```

Notice the comma. \
This is still a tuple.

## Selecting Multiple Columns

SELECT name, marks FROM students;

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()
cursor.execute("""
SELECT name, marks
FROM students
""")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

## fetchone()

Returns only the first row.

```python
cursor.execute("SELECT * FROM students")

student = cursor.fetchone()

print(student)
```

## fetchmany()

Returns a specified number of rows.

```python
cursor.execute("SELECT * FROM students")

students = cursor.fetchmany(2)

print(students)
```

### Comparison

| Method         | Returns  |
| -------------- | -------- |
| `fetchone()`   | One row  |
| `fetchmany(2)` | Two rows |
| `fetchall()`   | All rows |

## Mini Project

Display every student in a nice report.

```python
import sqlite3

connection = sqlite3.connect("college.db")

cursor = connection.cursor()

cursor.execute("SELECT * FROM students")

students = cursor.fetchall()

print("=" * 40)
print("      STUDENT REPORT")
print("=" * 40)

for student in students:

    print(f"ID    : {student[0]}")
    print(f"Name  : {student[1]}")
    print(f"Age   : {student[2]}")
    print(f"Marks : {student[3]}")

    print("-" * 40)

connection.close()
```

Q1: What is the purpose of the SELECT statement? \
Answer: It retrieves data from one or more database tables.

Q2: What is the difference between fetchone() and fetchall()? \
Answer: \
fetchone() returns a single row. \
fetchall() returns every matching row.

Q3: Why does SQLite return tuples? \
Answer: Each database row is represented as a Python tuple, making it easy to access columns by index.
