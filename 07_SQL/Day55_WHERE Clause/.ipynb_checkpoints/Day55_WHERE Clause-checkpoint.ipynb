# SQL WHERE Clause

## What is WHERE?

WHERE is used to filter records. \
It tells SQL: "Only show rows that match this condition."

### SQL Syntax

SELECT column_name \
FROM table_name \
WHERE condition;

General Format

SELECT * \
FROM students \
WHERE marks > 80;

Meaning: \
Show all columns \
From the students table \
But only where marks are greater than 80

## Comparison Operators

| Operator | Meaning               | Example         |
| -------- | --------------------- | --------------- |
| =        | Equal to              | `marks = 90`    |
| >        | Greater than          | `marks > 80`    |
| <        | Less than             | `marks < 80`    |
| >=       | Greater than or equal | `marks >= 90`   |
| <=       | Less than or equal    | `marks <= 60`   |
| !=       | Not equal             | `name != 'Ali'` |

## Example 1 – Marks Greater Than 80

SELECT * \
FROM students \
WHERE marks > 80;

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE marks > 80
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
INSERT INTO students (id, name, age, marks)
VALUES (4, 'Emma', 20, 95.5)
""")

cursor.execute("""
INSERT INTO students (id, name, age, marks)
VALUES (5, 'David', 22, 60.0)
""")

connection.commit()
connection.close()
```

## Example 2 – Marks Less Than 80

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE marks < 80
""")

students = cursor.fetchall()

for student in students:
    print(student)
```

## Example 3 – Find One Student by Name

SELECT * \
FROM students \
WHERE name = 'Ali';

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
WHERE name = ?
""", ("Ali",))

student = cursor.fetchall()

print(student)
```

## Why Use ? Instead of Writing the Name Directly?

Instead of:

```python
cursor.execute("SELECT * FROM students WHERE name='Ali'")
```

use:

```python
cursor.execute(
    "SELECT * FROM students WHERE name=?",
    ("Ali",)
)
```

Benefits
* Safer (prevents SQL Injection)
* Easier to change values
* Professional coding style

## Example 4 – Age Equal to 22

```python
cursor.execute("""
SELECT *
FROM students
WHERE age = 22
""")

students = cursor.fetchall()

for student in students:
    print(student)
```

## Using AND

Suppose you want: \
Marks greater than 80 \
Age greater than 20

SQL

SELECT * \
FROM students \
WHERE marks > 80 \
AND age > 20;

```python
cursor.execute("""
SELECT *
FROM students
WHERE marks > 80
AND age > 20
""")

students = cursor.fetchall()

for student in students:
    print(student)
```

Emma is not included because her age is 20.

## Using OR

Suppose you want: \
Students with marks above 90 \
OR students whose age is 23

SQL

SELECT * \
FROM students \
WHERE marks > 90 \
OR age = 23;

```python
cursor.execute("""
SELECT *
FROM students
WHERE marks > 90
OR age = 23;
""")

students = cursor.fetchall()

for student in students:
    print(student)
```

## AND vs OR

Think of a game.

**AND**

Both conditions must be true.

Marks > 80 \
AND \
Age > 20 \

Both must match.

**OR**

Only one condition needs to be true.

Marks > 90 \
OR \
Age = 23

Either one is enough.

## User Input Example

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

name = input("Enter student name: ")

cursor.execute("""
SELECT *
FROM students
WHERE name = ?
""", (name,))

student = cursor.fetchall()

print(student)

connection.close()
```

## Mini Project

Ask the user for minimum marks and display qualifying students.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

minimum_marks = float(input("Enter minimum marks: "))

cursor.execute("""
SELECT *
FROM students
WHERE marks >= ?
""", (minimum_marks,))

students = cursor.fetchall()

print("\nQualified Students")
print("-" * 40)

for student in students:
    print(f"ID    : {student[0]}")
    print(f"Name  : {student[1]}")
    print(f"Age   : {student[2]}")
    print(f"Marks : {student[3]}")
    print("-" * 40)

connection.close()
```

Note: SQLite accepts ==, but in standard SQL, = is the recommended operator. Learn and use = from the beginning.

Q1: What is the purpose of the WHERE clause? \
Answer: The WHERE clause filters rows based on a specified condition.

Q2: What is the difference between AND and OR? \
Answer: \
AND requires all conditions to be true. \
OR requires at least one condition to be true.

Q3: Why should parameterized queries (?) be used? \
Answer: They prevent SQL Injection, improve security, and make queries easier to maintain.
