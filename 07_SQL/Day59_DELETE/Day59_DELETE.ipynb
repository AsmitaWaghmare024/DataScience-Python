# SQL DELETE Statement

## What is DELETE?

DELETE removes rows from a table. \
Think: \
Database \
     │ \
 DELETE \
     │ \
Selected Rows Removed

Example:

| ID | Name | Marks |
| -- | ---- | ----: |
| 1  | Ali  |    90 |
| 2  | Sara |    85 |
| 3  | John |    78 |

John leaves the school. \
The teacher removes John's record from the register. \
Now the table becomes:

| ID | Name | Marks |
| -- | ---- | ----: |
| 1  | Ali  |    90 |
| 2  | Sara |    85 |

That is exactly what **DELETE** does.

### SQL Syntax
*DELETE FROM table_name \
WHERE condition;*

#### Golden Rule

Always remember: \
DELETE \
     ↓ \
FROM \
     ↓ \
WHERE

Without WHERE, every row will be deleted.

## Example 1 – Delete One Student

Delete David.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
DELETE FROM students
WHERE id = 5
""")

connection.commit()

print("Student Deleted Successfully!")

connection.close()
```

Verify the Deletion

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```
