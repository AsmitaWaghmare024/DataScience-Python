```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")

connection.commit()
connection.close()
```

```python
def add_student():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    marks = float(input("Enter Marks: "))

    cursor.execute("""
    INSERT INTO students
    VALUES(?,?,?,?)
    """,(id,name,age,marks))

    connection.commit()
    connection.close()

    print("\nStudent Added Successfully!\n")
```

```python
def view_students():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    cursor.execute("""
    SELECT *
    FROM students
    ORDER BY id
    """)

    students = cursor.fetchall()

    print("\n===== STUDENT LIST =====\n")

    for student in students:

        print(f"ID    : {student[0]}")
        print(f"Name  : {student[1]}")
        print(f"Age   : {student[2]}")
        print(f"Marks : {student[3]}")
        print("-"*30)

    connection.close()
```

```python
def search_student():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    name = input("Enter Student Name: ")

    cursor.execute("""
    SELECT *
    FROM students
    WHERE name = ?
    """,(name,))

    student = cursor.fetchone()

    if student:

        print("\nStudent Found\n")

        print(student)

    else:

        print("Student Not Found!")

    connection.close()
```

```python
def update_marks():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    student_id = int(input("Enter Student ID: "))
    new_marks = float(input("Enter New Marks: "))

    cursor.execute("""
    UPDATE students
    SET marks = ?
    WHERE id = ?
    """,(new_marks,student_id))

    connection.commit()

    print("Marks Updated!")

    connection.close()
```

```python
def delete_student():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    student_id = int(input("Enter Student ID: "))

    cursor.execute("""
    DELETE FROM students
    WHERE id = ?
    """,(student_id,))

    connection.commit()

    print("Student Deleted!")

    connection.close()
```

```python
def top_students():

    connection = sqlite3.connect("college.db")
    cursor = connection.cursor()

    cursor.execute("""
    SELECT *
    FROM students
    ORDER BY marks DESC
    LIMIT 3
    """)

    students = cursor.fetchall()

    print("\n===== TOP STUDENTS =====\n")

    rank = 1

    for student in students:

        print(f"Rank : {rank}")
        print(f"Name : {student[1]}")
        print(f"Marks: {student[3]}")
        print("-"*30)

        rank += 1

    connection.close()
```

```python
while True:

    print("\n===== STUDENT MANAGEMENT SYSTEM =====")

    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Marks")
    print("5. Delete Student")
    print("6. Top 3 Students")
    print("7. Exit")

    choice = input("\nEnter Choice: ")

    if choice == "1":

        add_student()

    elif choice == "2":

        view_students()

    elif choice == "3":

        search_student()

    elif choice == "4":

        update_marks()

    elif choice == "5":

        delete_student()

    elif choice == "6":

        top_students()

    elif choice == "7":

        print("Thank You!")

        break

    else:

        print("Invalid Choice!")
```

SQL Concepts Used

| SQL Command    | Used In           |
| -------------- | ----------------- |
| `CREATE TABLE` | Database creation |
| `INSERT`       | Add student       |
| `SELECT`       | View students     |
| `WHERE`        | Search student    |
| `ORDER BY`     | Sort students     |
| `LIMIT`        | Top 3 students    |
| `UPDATE`       | Update marks      |
| `DELETE`       | Remove student    |
