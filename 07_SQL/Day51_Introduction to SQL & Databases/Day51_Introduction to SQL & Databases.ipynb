## What is SQL?

SQL stands for: **Structured Query Language** \
It is the language used to talk to a database.

Example: \
Imagine asking: Show all students. \
SQL understands this request.

**Easy Memory**

## Primary Key (Introduction)

Every student should have a **unique ID**.

Example:

| ID | Name |
| -: | ---- |
|  1 | Ali  |
|  2 | Sara |
|  3 | John |

No two students should have the same ID.
This unique column is called the **Primary Key**.

## Using SQLite in Python

Python already includes SQLite. \
Import it:

```python
import sqlite3
```

```python
#Create a database:

import sqlite3

connection = sqlite3.connect("school.db")

print("Database Created Successfully!")
```

If **school.db** doesn't exist, Python creates it automatically.

connection = sqlite3.connect("school.db") Creates or opens the database named school.db.

## Closing the Database

Always close the connection.

```python
connection.close()

print("Database Closed")
```

```python
import sqlite3

connection = sqlite3.connect("school.db")

print("Database Created Successfully!")

connection.close()

print("Database Closed Successfully!")
```

Q1: What is SQL?

Answer:

SQL (Structured Query Language) is a language used to create, manage, and query relational databases.

Q2: What is a Database?

Answer:

A database is an organized collection of related data that can be stored, managed, and retrieved efficiently.

Q3: What is a DBMS?

Answer:

A Database Management System (DBMS) is software that allows users to create, store, manage, and manipulate databases.

Q4: Why is SQL important for Data Analysts?

Answer:

Because most business data is stored in databases, SQL allows analysts to retrieve, filter, summarize, and analyze data efficiently.
