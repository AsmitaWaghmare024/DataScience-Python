# SQL ORDER BY

## What is ORDER BY?
ORDER BY arranges rows in a specific order.

Messy Data \
     ↓ \
ORDER BY \
     ↓ \
Neatly Sorted Data

**Example:**

| Name  | Marks |
| ----- | ----: |
| Ali   |    90 |
| Sara  |    85 |
| John  |    78 |
| Emma  |    95 |
| David |    60 |

who got the highest marks? \
the list from highest to lowest:

| Name  | Marks |
| ----- | ----: |
| Emma  |    95 |
| Ali   |    90 |
| Sara  |    85 |
| John  |    78 |
| David |    60 |

This arranging is called **sorting**. \
In SQL, sorting is done using **ORDER BY**.

## SQL Syntax

SELECT column_name \
FROM table_name \
ORDER BY column_name;

## Ascending Order (ASC)

Ascending means: \
Small → Big \
A → Z \
1 → 100 \
ASC is the default order.

**Example**:

SELECT * \
FROM students \
ORDER BY marks ASC;

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks ASC
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## Descending Order (DESC)

Descending means: \
Big → Small \
Z → A \
100 → 1 

**Example:**

SELECT * \
FROM students \
ORDER BY marks DESC;

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks DESC
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## Sorting by Age

```python
connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT Name, age
FROM students
ORDER BY age ASC;
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## Sorting Multiple Columns
Suppose two students have the same age.

Sort by: 
1. Age
2. Then Name

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY age ASC, name ASC
""")

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

## Combining WHERE + ORDER BY
Show students scoring above 80, sorted from highest marks.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT Name, Marks
FROM students
WHERE marks > 80
ORDER BY marks DESC
""")

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

### User Input Example

Ask the user for the sort order.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

choice = input("Sort by marks (ASC/DESC): ").upper()

if choice == "DESC":
    query = "SELECT * FROM students ORDER BY marks DESC"
else:
    query = "SELECT * FROM students ORDER BY marks ASC"

cursor.execute(query)

students = cursor.fetchall()

for student in students:
    print(student)

connection.close()
```

## Mini Project

Display a ranked student report.

```python
import sqlite3

connection = sqlite3.connect("college.db")
cursor = connection.cursor()

cursor.execute("""
SELECT *
FROM students
ORDER BY marks DESC
""")

students = cursor.fetchall()

print("=" * 45)
print("        STUDENT RANK LIST")
print("=" * 45)

rank = 1

for student in students:

    print(f"Rank  : {rank}")
    print(f"Name  : {student[1]}")
    print(f"Age   : {student[2]}")
    print(f"Marks : {student[3]}")
    print("-" * 45)

    rank += 1

connection.close()
```

Q1: What is the purpose of ORDER BY? \
Answer: It sorts the query results in ascending or descending order.

Q2: What is the default sorting order? \
Answer: ASC (ascending).

Q3: What is the difference between ASC and DESC? \
Answer: \
ASC sorts from smallest to largest or A to Z. \
DESC sorts from largest to smallest or Z to A. 

Q4: Can ORDER BY sort using multiple columns? \
Answer: Yes. \ 
For example: \
ORDER BY age ASC, name ASC;

This sorts by age first, then by name for students with the same age.
