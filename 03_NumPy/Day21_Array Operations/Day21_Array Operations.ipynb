# Array Operations

Before NumPy, if you wanted to add 10 to every number:

```python
marks = [50, 60, 70]

new_marks = []

for mark in marks:
    new_marks.append(mark + 10)

print(new_marks)
```

Lots of work.

NumPy can do it in one line.

Imagine a classroom:

Student Marks

50, 60, 70

Teacher says 'Add 10 bonus marks to everyone'.

Instead of updating one by one:\
50 → 60\
60 → 70\
70 → 80

NumPy updates everyone together.

## Create an Array

```python
import numpy as np

marks = np.array([50, 60, 70])

print(marks)
```

## Add a Number to Entire Array

```python
marks = np.array([50,60,70])

print(marks + 10)
```

Every element gets +10.

## Subtract a Number

```python
marks = np.array([50,60,70])

print(marks - 5)
```

## Multiply a Number

```python
marks = np.array([50,60,70])

print(marks * 2)
```

## Divide a Number

```python
marks = np.array([50,60,70])

print(marks / 10)
```

Python List: [1,2,3] * 2

Output: [1,2,3,1,2,3]

Not multiplication.

```python
np.array([1,2,3]) * 2
```

## Array + Array

```python
# Suppose two students have marks from two tests.

test1 = np.array([50,60,70])

test2 = np.array([10,20,30])

# Add them

print(test1 + test2)
```

## Subtract Arrays

```python
a = np.array([100,200,300])

b = np.array([10,20,30])

print(a - b)
```

## Multiply Arrays

```python
a = np.array([1,2,3])

b = np.array([10,20,30])

print(a * b)
```

## Divide Arrays

```python
a = np.array([100,200,300])

b = np.array([10,20,30])

print(a / b)
```

## Power Operation

```python
arr = np.array([2,3,4])

print(arr ** 2)
```

## Square Root

```python
arr = np.array([4,9,16,25])

print(np.sqrt(arr))
```

## Sum of All Elements

```python
marks = np.array([50,60,70])

print(marks.sum())
```

## Average

```python
print(marks.mean())
```

## Maximum Value

```python
print(marks.max())
```

## Minimum Value

```python
print(marks.min())
```
