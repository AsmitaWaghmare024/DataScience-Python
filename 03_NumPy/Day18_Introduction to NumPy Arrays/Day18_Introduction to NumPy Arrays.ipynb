We already have lists.\

Scientists and data analysts need something faster.\

### NumPy is faster than Python lists?

Because it utilizes dense, contiguous, typed memory blocks executed via compiled C code, completely bypassing Python's heavy looping and object-checking overhead.

However, NumPy can become significantly slower than lists when performing element-by-element operations or managing small datasets due to fixed overhead of converting data between the C and python layers.

### Mini Project — Student Marks Analysis

```python
import numpy as np

marks = np.array([80, 65, 90, 75, 85])

print("Marks:", marks)

print("Total Students:", marks.size)

print("Average Marks:", marks.mean())

print("Highest Marks:", marks.max())

print("Lowest Marks:", marks.min())
```
