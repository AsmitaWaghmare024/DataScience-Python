# NumPy Reshaping Arrays

We'll learn how to change the shape of data without changing the data itself.

Think of LEGO blocks.

We have 6 blocks: 1 2 3 4 5 6

We can arrange them as:

1 2 3 \
4 5 6

or

1 2 \
3 4 \
5 6

Same blocks.

Different arrangement.

That's exactly what reshape() does.

## What is Reshaping?

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])

print(arr)
```

```python
print(arr.shape)
```

Meaning

1 Row \
6 Elements

## Reshape into 2 Rows and 3 Columns

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,3)

print(new_arr)
```

```python
print(new_arr.shape)
```

Before:

1 2 3 4 5 6

After:

1 2 3 \
4 5 6

### Another Example

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(3,2)

print(new_arr)
```

Shape: (3,2)

### Important Rule

Total elements must stay the same.

We have: 6 elements

Possible shapes:

2 × 3 = 6 ✅

3 × 2 = 6 ✅

1 × 6 = 6 ✅

6 × 1 = 6 ✅

Not possible:

4 × 2 = 8 ❌

### Error Example

```python
arr = np.array([1,2,3,4,5,6])

arr.reshape(4,2)
```

Because Need 8 values Have only 6 values

## Reshape into 1 Row

```python
arr = np.array([1,2,3,4])

new_arr = arr.reshape(1,4)

print(new_arr)
```

## Reshape into 1 Column

```python
arr = np.array([1,2,3,4])

new_arr = arr.reshape(4,1)

print(new_arr)
```

## Why Column Shape Matters?

Machine Learning often expects:

Age \
25 \
30 \
35 \
40

instead of: 25 30 35 40

So we'll frequently see:

```python
ages.reshape(-1,1)
```

## Using -1 (Automatic Calculation)

NumPy can calculate one dimension for us.

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,-1)

print(new_arr)
```

NumPy thinks:

2 Rows

Need total 6 elements

Columns = 3

### Example

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(-1,2)

print(new_arr)
```

NumPy calculates rows automatically.

## Flatten a 2D Array

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
```

```python
#Convert back to 1D

flat = arr.flatten()

print(flat)
```

```python
# Example

students = np.array([
    [80,90],
    [70,85],
    [95,88]
])

students.shape
```

2 Subjects
