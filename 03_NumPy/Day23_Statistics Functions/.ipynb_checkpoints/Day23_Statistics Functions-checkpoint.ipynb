# NumPy Statistics Function

Statistics = Understanding data using numbers

Example:

If a teacher has marks of 100 students, statistics helps answer:
* What is the average mark?
* Who scored highest?
* Who scored lowest?
* How spread out are the marks?

## Dataset

We'll use:

```python
import numpy as np

marks = np.array([80, 90, 70, 85, 95])
```

Student Marks

80 \
90 \
70 \
85 \
95

## Mean (Average)

The most common statistic.

Formula:   Mean =  Sum of all values / Number of values
    
Example

```python
print(marks.mean())
```

Manual Check \
80 + 90 + 70 + 85 + 95 = 420 \
420 ÷ 5 = 84 

Average = 84

## Sum

```python
print(marks.sum())
```

Meaning: Total marks of all students

## Maximum Value

Highest score.

```python
print(marks.max())
```

## Minimum Value

Lowest score.

```python
print(marks.min())
```

## Range

Range means: Highest - Lowest

Formula: Range = Maximum − Minimum

Example:

```python
marks.max() - marks.min()
```

## Median

Median = Middle value after sorting.

Example:

```python
marks = np.array([80,90,70,85,95])
np.median(marks)
```

## Why Median Matters

Suppose salaries:

```python
salaries = np.array([20000,25000,30000,35000,500000])
```

Average becomes very high because of one rich person.

Median gives a better picture of a "typical" salary.

Data scientists use median a lot.

## Standard Deviation (Easy Version)

Imagine two classes:

Class A: 50 51 49 50 50

Class B: 10 50 90 20 80

Both may have similar averages.

But:

Class A → Students are close together

Class B → Students are far apart

Standard deviation measures spread.

Code:

```python
marks.std()
```

### Easy Rule

Small Std → Data close together

Large Std → Data spread out

## Variance

Variance is related to standard deviation.

Formula: Standard Deviation=  under root (Variance)

Code:

```python
marks.var()
```

## Sorting Data

```python
np.sort(marks)
```

## Find Index of Maximum

```python
marks.argmax()
```

Largest value is at position 4

## Find Index of Minimum

```python
marks.argmin()
```

## Count Values

```python
len(marks)
```

```python
marks.size
```

## Statistics on 2D Arrays

Dataset:

```python
students = np.array([
    [80,90],
    [70,85],
    [95,88]
])
```

## Overall Average

```python
students.mean()
```

## Average of Each Column

```python
students.mean(axis=0)
```

Math Average \
Science Average

## Average of Each Row

```python
students.mean(axis=1)
```

Student 1 Average \
Student 2 Average \
Student 3 Average

## Understanding Axis

Remember:

axis=0 \
↓ \
Go Down \
Columns 

axis=1 \
→ \
Go Across \
Rows

data.mean() \

```python
data = np.array([
    [10,20],
    [30,40],
    [50,60]
])

print("Mean: ", data.mean())
print("Column: ",data.mean(axis=0))
print("Row: ",data.mean(axis=1))
```

## Mini Project 

Exam Result Analysis

```python
import numpy as np

marks = np.array([78,85,92,67,88,74,95])

print("Marks:", marks)

print("\nTotal:", marks.sum())

print("\nAverage:", marks.mean())

print("\nHighest:", marks.max())

print("\nLowest:", marks.min())

print("\nMedian:", np.median(marks))

print("\nStandard Deviation:", marks.std())

print("\nSorted Marks:", np.sort(marks))
```
