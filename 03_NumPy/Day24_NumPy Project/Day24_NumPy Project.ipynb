### Step 8: Passing Students

Suppose: Pass = Average ≥ 60

Calculate averages:

```python
student_avg = marks.mean(axis=1)

# Get passing students:

passed = student_avg >= 60

print(passed)
```

Only passing averages:

```python
print(student_avg[student_avg >= 60])
```

## Full Project Code

```python
import numpy as np

marks = np.array([
    [80,75,90],
    [60,70,65],
    [95,88,92],
    [50,55,60],
    [85,80,78]
])

print("Dataset:")
print(marks)

print("\nShape:")
print(marks.shape)

print("\nOverall Average:")
print(marks.mean())

print("\nSubject Averages:")
print(marks.mean(axis=0))

student_avg = marks.mean(axis=1)

print("\nStudent Averages:")
print(student_avg)

print("\nHighest Mark:")
print(marks.max())

print("\nLowest Mark:")
print(marks.min())

top_student = student_avg.argmax()

print("\nTop Student Index:")
print(top_student)

print("\nStudents Passed:")
print(student_avg[student_avg >= 60])

print("\nBonus Marks:")
print(marks + 5)

print("\nFlatten:")
print(marks.flatten())

print("\nSorted Marks:")
print(np.sort(marks.flatten()))
```
