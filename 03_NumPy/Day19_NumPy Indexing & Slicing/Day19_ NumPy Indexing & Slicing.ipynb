## Part 1: Indexing

Index means position.

Array

[10 20 30 40 50]

 0  1  2  3  4

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])

#First Element
print(arr[0])

#Second Element
print(arr[1])

#Third Element
print(arr[2])
```

### Negative Indexing

Python can count from the end.

[10 20 30 40 50]

-5 -4 -3 -2 -1

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[-1])
```

```python
print(arr[-2])
```

## Part 2: Slicing

Syntax:      arr[start:end]

Remember:
Start Included \
End Excluded 

```python
arr = np.array([10,20,30,40,50,60,70])
print(arr[1:4])
```

Why?

Position 1 → Included \
Position 4 → Excluded

```python
print(arr[0:3])
```

```python
print(arr[2:6])
```

### Beginning to End

```python
print(arr[:4])    #Python assumes: arr[0:4]
```

### Middle to End

```python
print(arr[3:])    #Python assumes: arr[3:last]
```

### Entire Array

```python
print(arr[:])
```

### Step Size

Syntax:

arr[start:end:step]

```python
#Every Second Number

print(arr[::2])
```

Explanation:

Take 1 \
Skip 1 \
Take 1 \
Skip 1

```python
# Every Third Number
print(arr[::3])
```

### Reverse Array

```python
print(arr[::-1])
```

### Modify Elements

NumPy arrays are mutable.

```python
#Change One Value

arr = np.array([10,20,30,40])

arr[1] = 99

print(arr)
```

```python
#Change Multiple Values

arr[1:4] = 100

print(arr)
```

NumPy updates multiple values at once.

### Fancy Indexing

```python
# Select multiple positions.

arr = np.array([10,20,30,40,50])

print(arr[[0,2,4]])
```

### Boolean Indexing

```python
marks = np.array([35,80,45,90,20])

#Get only passing marks.

print(marks > 40)
```

```python
print(marks[marks > 40])
```

Example

```python
ages = np.array([12,18,25,14,30,16,40])

print(ages[ages >= 18])
```

This is exactly how data cleaning works.

## Mini Project — Student Marks Filter

```python
import numpy as np

marks = np.array([35, 80, 60, 25, 95, 45])

print("All Marks:")
print(marks)

passed = marks[marks >= 40]

print("\nPassed Students:")
print(passed)

print("\nNumber Passed:")
print(len(passed))
```
