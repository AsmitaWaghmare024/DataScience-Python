# Three Types of Machine Learning
1. Supervised Learning
2. Unsupervised Learning
3. Reinforcement Learning

## Supervised Learning
Definition: The model learns from labeled data.

"Labeled" means we already know the correct answer.

| Hours Studied | Marks |
| ------------: | ----: |
|             2 |    35 |
|             3 |    45 |
|             5 |    60 |
|             8 |    95 |

Here,\
Input = Study Hours \
Output = Marks \
The computer learns: More study hours → Higher marks

**Example** \
Teacher gives: \
Question: 2 + 3 = ? \
Correct Answer: 5 \
The student learns because the correct answer is given. \
Machine Learning works the same way.

### Real-Life Applications
House Price Prediction \
Student Marks Prediction \
Email Spam Detection \
Disease Prediction \
Loan Approval \
Weather Prediction

Supervised Learning Diagram

Input Data \
      │ \
Correct Answers \
      │ \
      ▼ \
Machine Learning Model \
      │ \
      ▼ \
Prediction



```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4]])
y = np.array([20,40,60,80])

model = LinearRegression()

model.fit(X,y)

print(model.predict([[5]]))
```

### Types of Supervised Learning

There are 2 categories.

**A. Regression** \
Predicts numbers.

Examples: \
House Price \
Temperature \
Salary \
Marks

Example: \
Study Hours\
↓\
Marks = 85

**B. Classification** \
Predicts categories.

Eamples:
Spam / Not Spam \
Pass / Fail \
Cat / Dog \
Disease / No Disease 

Example: \
Email \
↓ \
Spam

Regression vs Classification
| Regression       | Classification      |
| ---------------- | ------------------- |
| Predicts numbers | Predicts categories |
| Salary           | Spam Detection      |
| Marks            | Pass/Fail           |
| House Price      | Cat/Dog             |

## Unsupervised Learning
Definition: The computer learns without correct answers.

No labels. \
It discovers patterns by itself.

**Example** \
You own a clothing store. \
Customers buy different clothes. \
You don't know their groups. \
Machine Learning automatically discovers:

Group A \
Jeans \
T-shirt \

Group B \
Suit \
Tie 

### Real-Life Applications
Customer Segmentation \
Market Basket Analysis \
Fraud Detection \
Product Recommendation \
Image Clustering

Example

| Name | Age |
| ---- | --: |
| Ali  |  20 |
| Sara |  21 |
| John |  35 |
| Emma |  36 |

The algorithm may create:

Group 1 \
Ali \
Sara 

Group 2 \
John \
Emma 

Nobody told it to do this.

```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1],[2],[8],[9]])

model = KMeans(n_clusters=2, random_state=42)

model.fit(X)

print(model.labels_)
```

Meaning: \
Group 0 → 1, 2 \
Group 1 → 8, 9
    
## Reinforcement Learning
Definition: The computer learns through rewards and penalties.

It tries different actions. \
Good action → Reward \
Bad action → Penalty

Dog Example
Sit \
↓ \
Reward  \
↓ \
Dog Learns Faster

### Real-Life Applications
Self-driving Cars \
Robots \
Chess AI \
Video Games \
Stock Trading \
Drone Navigation

### Reinforcement Learning Diagram

Environment \
      │ \
      ▼ \
Agent \
      │ \
Action \
      │ \
Reward/Penalty \
      │ \
Learn

**Example**

Robot tries: Move Left \
Hits Wall Penalty \
Robot remembers.

Next time: \
Move Right \
Reaches Goal \
Reward

## Comparison of All Three Types

| Feature         | Supervised       | Unsupervised    | Reinforcement     |
| --------------- | ---------------- | --------------- | ----------------- |
| Correct Answers |  Yes             |  No             |  No               |
| Learns From     | Labeled Data     | Unlabeled Data  | Rewards           |
| Goal            | Prediction       | Find Patterns   | Best Decision     |
| Example         | Marks Prediction | Customer Groups | Self-driving Cars |


Which Type Should I Use?

If I know the answer \
Use: Supervised Learning

If I want to discover hidden patterns \
Use: Unsupervised Learning

If the computer learns by trial and error \
Use: Reinforcement Learning

### Memory Trick

Teacher gives answers \
↓ \
Supervised

No teacher \
↓ \
Unsupervised

Reward & Punishment \
↓ \
Reinforcement

Q1. What are the three types of Machine Learning? \
Answer: \
Supervised Learning \
Unsupervised Learning \
Reinforcement Learning

Q2. What is Supervised Learning? \
Answer:It learns from labeled data where the correct output is already known.

Q3. What is Unsupervised Learning? \
Answer: It learns from unlabeled data and discovers hidden patterns or groups.

Q4. What is Reinforcement Learning? \
Answer: It learns through trial and error using rewards and penalties.

Q5. Difference between Regression and Classification? \
Answer: \
Regression: Predicts continuous numerical values (e.g., house prices, marks). \
Classification: Predicts categories or classes (e.g., spam/not spam, pass/fail).
