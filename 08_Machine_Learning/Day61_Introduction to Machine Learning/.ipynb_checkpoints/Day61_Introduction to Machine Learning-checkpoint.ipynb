# What is Machine Learning?
Machine Learning is a branch of Artificial Intelligence where computers learn patterns from data and make predictions or decisions without being explicitly programmed for every situation.

*Machine Learning = Learning from Data*

Example: \
Suppose I want to teach my younger brother to recognize apples.

* Method 1: Traditional Programming

tell him: \
Apples are round. \
Apples are red or green. \
Apples have a stem. \
Apples are fruits. \
I manually provide every rule.

* Method 2: Machine Learning

Instead, I show him: \
1000 apple pictures \
1000 banana pictures \
After seeing many examples, he learns the difference himself. \

This is exactly how Machine Learning works.

Machine Learning = Learning from Examples (Data).

## Why Do We Need Machine Learning?

Imagine I want to identify spam emails. \
You could write rules like: 

IF email contains "Win Money" \
THEN Spam

But spammers keep changing words. \
Instead, collect: \
100,000 spam emails \
100,000 normal emails

The ML model learns the patterns automatically.

## Traditional Programming vs Machine Learning

**Traditional Programming** \
Data  + Rules -> Answer

Example: \
*if marks >= 40:
    print("Pass")
else:
    print("Fail")*

**Machine Learning**\
Data + Correct Answers \
        ↓ \
Machine Learning Algorithm \
        ↓ \
Model \
        ↓ \
Predictions 

The computer learns the rules.

## Machine Learning Workflow
Collect Data \
      ↓ \
Clean Data \
      ↓ \
Split Data \
      ↓ \
Train Model \
      ↓ \
Test Model \
      ↓ \
Evaluate \
      ↓ \
Predict New Data


### Step 1: Collect Data
Example: Student data

| Hours | Marks |
| ----- | ----: |
| 2     |    35 |
| 3     |    45 |
| 5     |    60 |
| 8     |    95 |

### Step 2: Clean Data
Remove: \
Missing values \
Duplicate rows \
Wrong values \
We'll use Pandas for this

### Step 3: Split Data
Usually: \
80% → Training \
20% → Testing \ 
Why? \
We train the model on one part and test it on unseen data.

### Step 4: Train the Model
The algorithm studies the training data.

Example: \
Hours → Marks

It learns: \
More study hours usually lead to higher marks.

### Step 5: Test the Model
Give it new data. \
Example: Hours = 7 \
Prediction: Marks = 82

### Step 6: Evaluate
Check how accurate the predictions are.

### Step 7: Predict
Now the model is ready. \
You can give completely new data.

### Machine Learning Code

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Study hours
X = np.array([[2],
              [3],
              [5],
              [8]])

# Marks
y = np.array([35,45,60,95])

model = LinearRegression()

model.fit(X,y)

prediction = model.predict([[6]])

print("Predicted Marks:", prediction)
```

**What Happened?**

We gave the computer:

| Hours | Marks |
| ----- | ----: |
| 2     |    35 |
| 3     |    45 |
| 5     |    60 |
| 8     |    95 |

It learned the relationship. \
Then predicted: 6 Hours → 72 Marks

## Mini Project
Study Hours Predictor

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],
              [2],
              [3],
              [4],
              [5]])

y = np.array([30,40,50,60,70])

model = LinearRegression()

model.fit(X,y)

hours = float(input("Enter Study Hours: "))

prediction = model.predict([[hours]])

print(f"Predicted Marks: {prediction[0]:.2f}")
```
