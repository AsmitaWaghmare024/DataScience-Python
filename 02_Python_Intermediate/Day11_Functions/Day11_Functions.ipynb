# Return vs Print

Many beginners confuse these.

```python
def add(a, b):
    print(a + b)

result = add(5, 3)

print(result)
```

Why?

Because print() only displays.

It doesn't send a value back.

# Using Return

```python
def add(a, b):
    return a + b

result = add(5, 4 )

print(result)
```

Think Like a Delivery Service
print()

|

Shows package

return

|

Delivers package back

# Keyword Arguments

```python
#Normal:

def student(name, age):
    print(name, age)

student("James", 24)
```

```python
#Keyword version:

student(age=24, name="James")
```

# Variable Scope

Global Variable

```python
name = "James"

def show():
    print(name)

show()

#Function can see global variables.
```

Local Variable

```python
def show():
    age = 24
    print(age)

show()
#Works.       But
```

```python
print(age)
```

Error 

Because age exists only inside the function.

# Multiple Returns

```python
def calculations(a, b):
    return a+b, a-b

result = calculations(10, 5)

print(result)
```

# Unpacking

```python
sum_result, diff_result = calculations(10, 5)

print(sum_result)
print(diff_result)
```

Mini Project

Student Grade System

```python
def calculate_grade(mark):

    if mark >= 90:
        return "A"

    elif mark >= 80:
        return "B"

    elif mark >= 70:
        return "C"

    else:
        return "Fail"


student_mark = int(input("Enter mark: "))

print("Grade:", calculate_grade(student_mark))
```
