* Creating Objects

```python
class Student:

    def __init__(self, name, age):

        self.name = name
        self.age = age

s1 = Student("James", 24)

print(s1.name)
print(s1.age)
```

* Concept 4: Methods

Methods are functions inside classes.

Example:

```python
class Student:

    def __init__(self, name):

        self.name = name

    def say_hello(self):

        print("Hello", self.name)
        
# using method

s1 = Student("James")

s1.say_hello()
```

Class: Student

Method: say_hello()

Object: James

Result: Hello James

Example 2: Dog Class

```python
class Dog:

    def __init__(self, name):

        self.name = name

    def bark(self):

        print(self.name, "says Woof!")
        
#create

d1 = Dog("Tommy")

d1.bark()
```

Example 3: Car Class

```python
class Car:

    def __init__(self, brand):

        self.brand = brand

    def show(self):

        print("Car Brand:", self.brand)
        
#create

c1 = Car("BMW")

c1.show()
```

Example 4: Student Marks

```python
class Student:

    def __init__(self, name, marks):

        self.name = name
        self.marks = marks

    def display(self):

        print(self.name)
        print(self.marks)
        
#use
s1 = Student("James", 89)

s1.display()
```

Multiple Objects\
s1 = Student("James", 89)\
s2 = Student("Ali", 90)\
s3 = Student("John", 70)\

Each object stores different values.

Mini Project 1 — Student Information System

```python
class Student:

    def __init__(self, name, marks):

        self.name = name
        self.marks = marks

    def display(self):

        print("Name:", self.name)
        print("Marks:", self.marks)
        
s1 = Student("James", 85)

s1.display()
```

Mini Project 2 — Bank Account

```python
class BankAccount:

    def __init__(self, owner, balance):

        self.owner = owner
        self.balance = balance

    def show_balance(self):

        print("Owner:", self.owner)
        print("Balance:", self.balance)
        
acc1 = BankAccount("James", 5000)

acc1.show_balance()
```
