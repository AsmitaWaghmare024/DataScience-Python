# Filter

Think of filter() as a security guard.

It looks at every item and decides:

>Keep it 

>Remove it 

Imagine a school gate.

The security guard checks every student:

Student 1 → Pass? → YES → Enter
Student 2 → Pass? → NO → Stay outside
Student 3 → Pass? → YES → Enter

filter() works exactly like that.

It checks every item and keeps only the ones that satisfy the condition.

Example 1 

Keep Passing Marks

```python
#We only want marks ≥ 40.

marks = [25, 40, 55, 30, 80]

passed = list(filter(lambda x: x >= 40, marks))

print(passed)
```

Python checks:

25 >= 40 ?  No\
40 >= 40 ?  Yes\
55 >= 40 ?  Yes\
30 >= 40 ?  No\
80 >= 40 ?  Yes

Example 2 

Keep Even Numbers

```python
numbers = [1,2,3,4,5,6,7,8]

evens = list(filter(lambda x: x % 2 == 0, numbers))  #x % 2 == 0 means No remainder when divided by 2

print(evens)
```

Example 3 

Keep Odd Numbers

```python
numbers = [1,2,3,4,5,6,7,8]

odds = list(filter(lambda x: x % 2 != 0, numbers))

print(odds)
```

Example 4 

Adults Only

```python
ages = [12, 18, 25, 14, 30]

adults = list(filter(lambda age: age >= 18, ages))

print(adults)
```

Example 5 

Positive Numbers

```python
numbers = [-5, 10, -2, 7, -1, 20]

positive = list(filter(lambda x: x > 0, numbers))

print(positive)
```

Example 6 

Long Names

```python
names = ["Ali", "James", "John", "mily", "Li"]

long_names = list(filter(lambda name: len(name) > 3, names))

print(long_names)
```

Using a Normal Function

We don't always need lambda.

```python
def is_pass(mark):
    return mark >= 40

marks = [25,40,55,30,80]

passed = list(filter(is_pass, marks))

print(passed)
```

map() vs filter()

```python
# map() Changes every item.

numbers = [1,2,3]

result = list(map(lambda x: x*2, numbers))

print(result)

#Every number changed.
```

```python
# filter() Keeps only some items.

numbers = [1,2,3,4,5]

result = list(filter(lambda x: x%2==0, numbers))

print(result)

#Some numbers removed.
```

Easy Memory Trick

map()\
↓\
MODIFIES data\
\
filter()\
↓\
REMOVES unwanted data
