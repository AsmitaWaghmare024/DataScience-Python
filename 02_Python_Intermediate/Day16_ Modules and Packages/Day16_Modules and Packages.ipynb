> What is a Module?\
A module is simply 'A Python file (.py)'

Example:\
File: calculator.py

def add(a, b):\
    return a + b

def multiply(a, b):\
    return a * b

#This file is now a module.

* Using the Module

Main file:

```python
import calculator

print(calculator.add(5, 3))
print(calculator.multiply(4, 2))
```

* Import Specific Function

```python
#instead of:

import math

print(math.sqrt(25))
```

```python
# we can do
from math import sqrt

print(sqrt(25))
```

* Import Multiple Functions

```python
from math import sqrt, pi

print(sqrt(49))
print(pi)
```

* Rename a Module

Sometimes names are long.

```python
import math as m

print(m.sqrt(81))
```

Why Aliases Matter

* What is a Package?

A package is A folder containing modules

Package
│\
├── module1.py\
├── module2.py\
├── module3.py

Example:
    
NumPy Package
│\
├── arrays\
├── math tools\
├── statistics tools

Pandas Package
│\
├── DataFrames\
├── CSV tools\
├── Excel tools

Module vs Package

| Module             | Package          |
| ------------------ | ---------------- |
| Single .py file    | Folder           |
| Contains functions | Contains modules |
| Small              | Bigger           |

Mini Project — Random Password Generator

```python
import random

characters = "abcdefghijklmnopqrstuvwxyz"

password = ""

for i in range(6):
    password += random.choice(characters)

print("Password:", password)
```

Every run gives a different password.
