# List Comprehensions

Suppose we want squares of numbers.

Normal way:

```python
squares = []

for i in range(1, 6):
    squares.append(i * i)

print(squares)
```

Works perfectly.

But Python gives us a shorter way.

List Comprehension

Same thing:

```python
squares = [i * i for i in range(1, 6)]

print(squares)
```

Exactly the same result.

Structure

[new_value for item in collection]

```python
numbers = [1, 2, 3]

double = [n * 2 for n in numbers]

print(double)
```

numbers

├── 1 → 2

├── 2 → 4

└── 3 → 6

Convert Names to Uppercase

```python
#Normal way:

names = ["rahul", "john", "ali"]

result = []

for name in names:
    result.append(name.upper())

print(result)
```

```python
#List comprehension:

names = ["rahul", "john", "ali"]

result = [name.upper() for name in names]

print(result)
```

Even Numbers

Get even numbers from 1–10.

```python
evens = [n for n in range(1, 11) if n % 2 == 0]   # n for n in range(1,11): means Take every number  
                                                  # if n % 2 == 0 means Only keep even numbers
print(evens)
```

Odd Numbers

```python
odds = [n for n in range(1,11) if n % 2 != 0]

print(odds)
```

Length of Words

```python
words = ["apple", "banana", "mango"]

lengths = [len(word) for word in words]

print(lengths)
```

```python
#Customer names
customers = ["rahul", "john", "ali"]

#Clean them:

cleaned = [name.capitalize() for name in customers]

print(cleaned)
```

This kind of cleaning happens all the time in datasets.

Nested List Comprehension (Easy Introduction)

Suppose:

```python
matrix = [
    [1,2],
    [3,4]
]

#Get all values

numbers = [num for row in matrix for num in row]

print(numbers)
```

Don't worry if this looks strange. Just know it exists.
