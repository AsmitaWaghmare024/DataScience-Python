# Map Function

Learn how to apply the same operation to every item in a list.

Think of map() as a magic machine.

Input List -> Map Machine -> New List

First Understand the Problem

```python
marks = [10, 20, 30, 40]

#Teacher gives everyone +5 marks. Without map():

new_marks = []

for mark in marks:
    new_marks.append(mark + 5)

print(new_marks)
```

Works fine.

But Python gives shorter way.

First map()

```python
marks = [10, 20, 30, 40]

new_marks = list(map(lambda x: x + 5, marks))

print(new_marks)
```

Look at:

map(lambda x: x + 5, marks)

Python does:

10 -> 15
20 -> 25
30 -> 35
40 -> 45

Then returns all results.

Why Use list()?

map() creates a special map object.

To see actual values:

list(...)

converts it into a list.

Double Numbers

```python
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, numbers))

print(result)
```

Square Numbers

```python
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x * x, numbers))

print(squares)
```

Convert Names to Uppercase

```python
names = ["James", "john", "ali"]

result = list(map(lambda name: name.upper(), names))

print(result)
```

Get Length of Words

```python
words = ["apple", "banana", "mango"]

lengths = list(map(len, words))

print(lengths)
```

Using a Normal Function with map()

```python
#You don't always need lambda.

def square(x):
    return x * x

numbers = [1, 2, 3, 4]

result = list(map(square, numbers))

print(result)
```

```python
prices = [100, 200, 300, 400]

#Add 18% tax:

prices_with_tax = list(
    map(lambda price: price * 1.18, prices)
)

print(prices_with_tax)
```

This kind of transformation happens constantly in data analysis.

Mini Project

Temperature Converter

```python
#Suppose temperatures are in Celsius:
celsius = [0, 20, 30, 40]

#Convert to Fahrenheit.

fahrenheit = list(
    map(lambda c: (c * 9/5) + 32, celsius)
)

print(fahrenheit)
```

When Should we Use map()?

Use map() when we want to change EVERY item.

Examples:

1. Add tax
2. Add bonus marks
3. Convert lowercase → uppercase
4. Convert Celsius → Fahrenheit
5. Calculate squares

When NOT to Use map()

If we want to remove some items:

> Keep only even numbers

> Keep only passed students

> Keep only adults

Use:

filter()
