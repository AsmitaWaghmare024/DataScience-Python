# Sorting (sort() and sorted())

Imagine This Situation
\
You have marks: marks = [85, 60, 95, 70]
\
They are mixed up.
\
We want: [60, 70, 85, 95]
\
Python can sort them.

Method 1: sorted()

```python
marks = [85, 60, 95, 70]

result = sorted(marks)

print(result)
```

print(marks)
\
still gives: [85, 60, 95, 70]
\
Because:
\
sorted()
creates a NEW sorted list
\
Original list remains unchanged.

Method 2: sort()

```python
marks = [85, 60, 95, 70]

marks.sort()

print(marks)
```

# Difference

sorted()\
new_list = sorted(marks)
\
Creates a new list.

sort()\
marks.sort()
\
Changes the original list.

Ascending Order

```python
numbers = [5, 2, 8, 1, 9]

print(sorted(numbers))
```

Descending Order

```python
numbers = [5, 2, 8, 1, 9]

print(sorted(numbers, reverse=True))
```

Sort Names

```python
names = ["James", "Ali", "John", "Amit"]

print(sorted(names))
```

Sort by Length

```python
names = ["James", "Ali", "Christopher", "Tom"]

result = sorted(names, key=len)

print(result)
```

 * Understanding key

Python asks: How should I compare items?\
answer:key=len\

Meaning:Compare using length

Lambda + Sorting

```python
names = ["Rahul", "Ali", "Christopher", "Tom"]

result = sorted(
    names,
    key=lambda x: len(x)
)

print(result)
```

Employee salaries:

```python
salaries = [45000, 70000, 50000, 90000]

result = sorted(
    salaries,
    reverse=True
)

print(result)
```

Sorting Tuples

```python
students = [
    ("Ali", 85),
    ("Rahul", 95),
    ("John", 70)
]

result = sorted(
    students,
    key=lambda x: x[1]    lambda x: x[1]   #means Sort using marks
)

print(result)
```

Sort Highest Marks First

```python
result = sorted(
    students,
    key=lambda x: x[1],
    reverse=True
)

print(result)
```

Mini Project \
Student Ranking System

```python
students = [
    ("James", 92),
    ("Ali", 85),
    ("John", 78),
    ("Amit", 95)
]

ranking = sorted(
    students,
    key=lambda x: x[1],
    reverse=True
)

print("Student Ranking")

for rank, student in enumerate(ranking, start=1):
    print(rank, student)
```
