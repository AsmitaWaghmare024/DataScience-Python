# Lambda Functions


A lambda function is simply:

A small function written in one line.

```python
# First Understand the Problem

# Normal function:

def square(x):
    return x * x

print(square(5))
```

Works perfectly.

But if we need a very small function that we'll use only once, writing 3 lines feels unnecessary.

Lambda Solution

```python
square = lambda x: x * x

print(square(5))
```

Same result.

Example 1

```python
double = lambda x: x * 2

print(double(10))
```

Example 2

```python
# Add two numbers:

add = lambda a, b: a + b

print(add(5, 3))
```

Check Even Number

```python
is_even = lambda n: n % 2 == 0

print(is_even(8))
```

Get First Letter

```python
first_letter = lambda name: name[0]

print(first_letter("James"))
```

Sorting Example

```python
students = [
    ("Ali", 85),
    ("James", 95),
    ("John", 70)
]

students.sort(key=lambda x: x[1])

print(students)
```

Lambda + List Comprehension

```python
numbers = [1,2,3,4]

square = lambda x: x*x

result = [square(n) for n in numbers]

print(result)
```

When NOT to Use Lambda

If logic becomes large:

```python
def calculate_grade(mark):

    if mark >= 90:
        return "A"

    elif mark >= 80:
        return "B"

    else:
        return "Fail"
```

This should remain a normal function.

Lambda is for short operations.

Student Pass Checker

```python
check_pass = lambda mark: "Pass" if mark >= 40 else "Fail"

print(check_pass(55))
print(check_pass(30))
print(check_pass(70))
```
