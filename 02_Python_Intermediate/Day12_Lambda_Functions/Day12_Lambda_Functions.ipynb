# Lambda Functions

A lambda function is simply "A small function written in one line.""

First Understand the Problem

```python
#Normal function:

def square(x):
    return x * x

print(square(5))
```

Works perfectly.

But if we need a very small function that we'll use only once, writing 3 lines feels unnecessary.

```python
#Lambda Solution

square = lambda x: x * x

print(square(5))
```

Imagine:

Normal Function Build a full shop 

Lambda Function Use a small food cart 

Both sell food.

One is much smaller.

```python
double = lambda x: x * 2

print(double(10))
```

```python
add = lambda a, b: a + b

print(add(5, 3))
```

Compare

#Lambda:

lambda a, b: a + b

Same logic.

Check Even Number

```python
is_even = lambda n: n % 2 == 0

print(is_even(8))
```

Get First Letter

```python
first_letter = lambda name: name[0]

print(first_letter("James"))
```

It works beautifully with:

1. sorting
2. filtering
3. Pandas
4. Machine Learning preprocessing

Sorting Example

```python
#Sort by marks

students = [
    ("Ali", 85),
    ("James", 95),
    ("John", 70)
]

students.sort(key=lambda x: x[1])

print(students)
```

Lambda + List Comprehension

```python
numbers = [1,2,3,4]

square = lambda x: x * x

result = [square(n) for n in numbers]

print(result)
```

Common Mistakes
